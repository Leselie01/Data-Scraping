```python
<div id="kakaoWrap" class="newsview_media"> 
   <div id="wrapMinidaum"><div id="minidaum"><strong class="minidaum_logo"><a href="https://www.daum.net/?nil_profile=mini&amp;nil_src=daum" target="_top" class="img_logo minidaum_tiara_click" data-tiara-layer="minidaum logo" data-tiara-action-name="미니다음_로고">Daum</a></strong><div class="minidaum_gnb"><h2 class="screen_out">서비스 바로가기</h2><div id="minidaumUser"><a href="https://logins.daum.net/accounts/loginform.do?url=https%3A%2F%2Fnews.v.daum.net%2Fv%2F20220227224906021" target="_top" id="btnMinidaumLogin" class="minidaum_login minidaum_tiara_click" data-tiara-layer="minidaum login" data-tiara-action-name="미니다음_로고">로그인</a></div><div id="minidaumService"><div class="dsvc_util"><a href="https://mail.daum.net/?nil_profile=mini&amp;nil_src=mail" id="minidaumMail" class="minidaum_mail minidaum_tiara_click" data-tiara-layer="minidaum mail" data-tiara-action-name="미니다음_메일"><span class="img_minidaum ico_dsvc_mail">메일</span><em class="minidaum_count" id="minidaumMailCntWrapper"><span class="screen_out">받은메일수</span><span id="minidaumMailCnt"></span></em></a><div id="minidaumMailLayer" class="minidaum_layer"><span class="img_minidaum ico_arrow"></span><div class="inner_minidaum_layer"><span class="img_minidaum ico_dsvc_notify"></span><p class="minidaum_notify"><em>메일함이 없습니다</em><br>메일함 생성을 위해선 Daum아이디가 필요합니다</p><a href="https://accounts.daum.net/mail/create" class="minidaum_account">Daum아이디 만들기</a></div></div></div><div class="dsvc_util"><a href="http://www.daum.net/doc/sitemap/?nil_profile=mini&amp;nil_src=sitemap" target="_top" class="minidaum_sitemap" role="button" aria-expanded="false" id="minidaumMore"><span class="img_minidaum ico_dsvc_sitemap">Daum 미니 사이트맵</span></a><div id="minidaumMoreLayer" class="minidaum_layer"><span class="img_minidaum ico_arrow"></span><div class="inner_minidaum_layer"><strong class="screen_out">주요서비스 바로가기</strong><ul class="minidaum_list_dsvc"><li><a href="https://cafe.daum.net/?nil_profile=mini&amp;nil_src=cafe" target="_top" class="minidaum_tiara_click" data-tiara-layer="minidaum site_layer cafe" data-tiara-action-name="미니다음_서비스 보기 레이어_서비스명 클릭"><span class="dsvc_thumb"><img src="//t1.daumcdn.net/daumtop_deco/images/common/minidaum/rtn/ic_cafe.png" width="36" height="36" alt=""></span><span class="dsvc_txt">카페</span></a></li><li><a href="https://news.daum.net/?nil_profile=mini&amp;nil_src=news" target="_top" class="minidaum_tiara_click" data-tiara-layer="minidaum site_layer news" data-tiara-action-name="미니다음_서비스 보기 레이어_서비스명 클릭"><img src="//t1.daumcdn.net/daumtop_deco/images/common/minidaum/rtn/ic_news.png" width="36" height="36" alt=""><span class="dsvc_txt">뉴스</span></a></li><li><a href="https://map.kakao.com/?nil_profile=mini&amp;nil_src=map" class="minidaum_tiara_click" data-tiara-layer="minidaum site_layer map" data-tiara-action-name="미니다음_서비스 보기 레이어_서비스명 클릭"><span class="dsvc_thumb"><img src="//t1.daumcdn.net/daumtop_deco/images/common/minidaum/rtn/ic_map.png" width="36" height="36" alt=""></span><span class="dsvc_txt">지도</span></a></li><li><a href="https://shoppinghow.kakao.com/top?nil_profile=mini&amp;nil_src=shophow" class="minidaum_tiara_click" data-tiara-layer="minidaum site_layer shopping" data-tiara-action-name="미니다음_서비스 보기 레이어_서비스명 클릭"><span class="dsvc_thumb"><img src="//t1.daumcdn.net/daumtop_deco/images/common/minidaum/rtn/ic_shopping.png" width="36" height="36" alt=""></span><span class="dsvc_txt">쇼핑</span></a></li></ul><div class="minidaum_listbox"><strong class="screen_out">서비스 바로가기</strong><strong class="minidaum_idx minidaum_idx1">더보기 1열</strong><ul class="minidaum_list_group minidaum_list_group1"><li><a href="https://blog.daum.net/?nil_profile=mini&amp;nil_src=blog" target="_top" data-tiara-layer="minidaum site_layer blog" data-tiara-action-name="미니다음_서비스 보기 레이어_서비스명 클릭">블로그</a></li><li><a href="https://www.tistory.com/?nil_profile=mini&amp;nil_src=tistory" target="_top" data-tiara-layer="minidaum site_layer tistory" data-tiara-action-name="미니다음_서비스 보기 레이어_서비스명 클릭">티스토리</a></li><li><a href="https://brunch.co.kr/?nil_profile=mini&amp;nil_src=brunch" target="_top" data-tiara-layer="minidaum site_layer brunch" data-tiara-action-name="미니다음_서비스 보기 레이어_서비스명 클릭">브런치</a></li><li><a href="https://tv.kakao.com/?nil_profile=mini&amp;nil_src=kakaotv" target="_top" data-tiara-layer="minidaum site_layer kakaotv" data-tiara-action-name="미니다음_서비스 보기 레이어_서비스명 클릭">카카오TV</a></li><li><a href="https://game.daum.net/?nil_profile=mini&amp;nil_src=game" target="_top" data-tiara-layer="minidaum site_layer game" data-tiara-action-name="미니다음_서비스 보기 레이어_서비스명 클릭">게임</a></li></ul><strong class="minidaum_idx minidaum_idx2">더보기 2열</strong><ul class="minidaum_list_group minidaum_list_group2"><li><a href="https://sports.daum.net/?nil_profile=mini&amp;nil_src=sports" target="_top" data-tiara-layer="minidaum site_layer sports" data-tiara-action-name="미니다음_서비스 보기 레이어_서비스명 클릭">스포츠</a></li><li><a href="https://entertain.daum.net/?nil_profile=mini&amp;nil_src=entertain" target="_top" data-tiara-layer="minidaum site_layer entertain" data-tiara-action-name="미니다음_서비스 보기 레이어_서비스명 클릭">연예</a></li><li><a href="https://movie.daum.net/main" target="_top" data-tiara-layer="minidaum site_layer movie" data-tiara-action-name="미니다음_서비스 보기 레이어_서비스명 클릭">영화</a></li><li><a href="https://webtoon.kakao.com/?nil_profile=mini&amp;nil_src=webtoon" target="_top" data-tiara-layer="minidaum site_layer webtoon" data-tiara-action-name="미니다음_서비스 보기 레이어_서비스명 클릭">웹툰</a></li><li><a href="https://www.melon.com/?nil_profile=mini&amp;nil_src=melon" target="_top" data-tiara-layer="minidaum site_layer melon" data-tiara-action-name="미니다음_서비스 보기 레이어_서비스명 클릭">멜론</a></li></ul><strong class="minidaum_idx minidaum_idx3">더보기 3열</strong><ul class="minidaum_list_group minidaum_list_group3"><li><a href="https://dic.daum.net/?nil_profile=mini&amp;nil_src=dic" target="_top" data-tiara-layer="minidaum site_layer dic" data-tiara-action-name="미니다음_서비스 보기 레이어_서비스명 클릭">어학사전</a></li><li><a href="https://translate.kakao.com/" target="_top" data-tiara-layer="minidaum site_layer translate" data-tiara-action-name="미니다음_서비스 보기 레이어_서비스명 클릭">번역</a></li><li><a href="https://wordbook.daum.net/my/search/log.do?dic_type=endic" target="_top" data-tiara-layer="minidaum site_layer wordbook" data-tiara-action-name="미니다음_서비스 보기 레이어_서비스명 클릭">단어장</a></li><li><a href="https://dic.daum.net/grammar_checker.do" target="_top" data-tiara-layer="minidaum site_layer grammar_checker" data-tiara-action-name="미니다음_서비스 보기 레이어_서비스명 클릭">맞춤법검사</a></li><li><a href="https://100.daum.net/" target="_top" data-tiara-layer="minidaum site_layer 100" data-tiara-action-name="미니다음_서비스 보기 레이어_서비스명 클릭">백과사전</a></li></ul><strong class="minidaum_idx minidaum_idx4">더보기 4열</strong><ul class="minidaum_list_group minidaum_list_group4"><li><a href="https://finance.daum.net/" target="_top" data-tiara-layer="minidaum site_layer finance" data-tiara-action-name="미니다음_서비스 보기 레이어_서비스명 클릭">금융</a></li><li><a href="https://realty.daum.net/" target="_top" data-tiara-layer="minidaum site_layer realty" data-tiara-action-name="미니다음_서비스 보기 레이어_서비스명 클릭">부동산</a></li><li><a href="https://auto.daum.net/" target="_top" data-tiara-layer="minidaum site_layer auto" data-tiara-action-name="미니다음_서비스 보기 레이어_서비스명 클릭">자동차</a></li><li><a href="https://makers.kakao.com/" target="_top" data-tiara-layer="minidaum site_layer makers" data-tiara-action-name="미니다음_서비스 보기 레이어_서비스명 클릭">메이커스</a></li><li><a href="https://together.kakao.com/?nil_profile=mini&amp;nil_src=together" target="_top" data-tiara-layer="minidaum site_layer together" data-tiara-action-name="미니다음_서비스 보기 레이어_서비스명 클릭">같이가치</a></li><li class="last minidaum_all"><a href="https://www.daum.net/doc/sitemap/?nil_profile=mini&amp;nil_src=all" target="_top" data-tiara-layer="minidaum site_layer sitemap" data-tiara-action-name="미니다음_서비스 보기 레이어_전체보기 클릭">전체보기</a></li></ul></div></div></div></div></div></div></div></div> 
   <div id="kakaoHead" role="banner" data-cloud-area="header"> 
    <div class="head_media" data-cloud="pc_header_gnb"> 
     <h1> <a href="https://www.daum.net" class="link_daum"> <img src="//t1.daumcdn.net/media/news/news2016/retina/logo_daum.jpg" width="45" height="19" class="thumb_g" alt="Daum"> </a> <a href="//news.daum.net" id="kakaoServiceLogo"><span class="ir_wa">뉴스</span></a> </h1> <strong class="screen_out">관련서비스</strong> 
     <ul id="kakaoRelServices"> 
      <li><a href="https://entertain.daum.net/" class="link_services link_services1"><span class="ir_wa">연예</span></a></li> 
      <li><a href="https://sports.daum.net/" class="link_services link_services2"><span class="ir_wa">스포츠</span></a></li> 
     </ul> 
     <h2 class="screen_out">검색</h2> 
     <div class="search_news" id="search_news"> 
      <form id="kakaoSearch" name="kakaoSearch" action="https://search.daum.net/search" method="get"> 
       <fieldset class="fld_inside"> <legend class="screen_out">검색어 입력폼</legend> 
        <div class="box_searchbar"> <label for="qVal" class="lab_search">통합검색</label> 
         <input type="hidden" name="nil_profile" value="search"> 
         <input type="hidden" name="nil_src" value="media"> 
         <input type="hidden" name="DA" value="23A"> 
         <input type="hidden" name="rtmaxcoll" value="NNS"> 
         <input type="text" class="tf_keyword" id="qVal" name="q" value="" title="검색어 입력" autocomplete="off" spellcheck="false" data-bind="value: keywordInput.searchQuery, hasFocus: keywordInput.isSelected, event: { keydown: onKeyDown, keyup: onKeyUp, focus: onFocus, blur: onBlur }"><span aria-live="polite" data-bind="text: suggestList.accessible, attr: { 'aria-live': 'polite'}, css: 'screen_out'" class="screen_out"></span> <button id="kakaoBtnSearch" class="btn_search" type="submit"><span class="ir_wa">검색</span></button> 
        </div> 
       </fieldset> 
       <div id="wrap_suggestBox" class="wrap_suggestBox"> 
        <div id="suggestBox" class="suggestBox box_sub" data-bind="template: { name: &quot;suggest-container&quot; }"><div class="baseBox bg" data-bind="
	visible: (suggestList.length() > 0 &amp;&amp; suggestList.visible) || suggestList.enableAnswerLayer,
	template: { name: 'suggest-layer' }" style="display: none;"><!-- ko if: options.isReverse --><!-- /ko -->

<!-- ko ifnot: options.isReverse -->
    <div data-bind="foreach: suggestList.lists"></div>

    <!-- ko if: options.enableFooter --><!-- /ko -->
<!-- /ko -->
</div>
<!-- ko if: options.enableRecentList --><!-- /ko -->
</div> 
       </div> 
      </form> 
     </div> 
    </div> 
    <div id="kakaoGnb" role="navigation" data-cloud="pc_header_gnb"> 
     <div class="inner_gnb"> 
      <h2 class="screen_out">뉴스 메인메뉴</h2> 
      <ul class="gnb_comm" data-category="economic"> 
       <li class="home"><a href="//news.daum.net/" class="link_gnb link_gnb1"><span class="ir_wa">홈</span><span class="bar_gnb"><span class="inner_bar"></span></span></a></li> 
       <li class="society"><a href="//news.daum.net/society/" class="link_gnb link_gnb2"><span class="ir_wa">사회</span><span class="bar_gnb"><span class="inner_bar"></span></span></a></li> 
       <li class="politics"><a href="//news.daum.net/politics/" class="link_gnb link_gnb3"><span class="ir_wa">정치</span><span class="bar_gnb"><span class="inner_bar"></span></span></a></li> 
       <li class="economic on"><a href="//news.daum.net/economic/" class="link_gnb link_gnb4"><span class="ir_wa">경제</span><span class="bar_gnb"><span class="inner_bar"></span></span></a></li> 
       <li class="foreign"><a href="//news.daum.net/foreign/" class="link_gnb link_gnb5"><span class="ir_wa">국제</span><span class="bar_gnb"><span class="inner_bar"></span></span></a></li> 
       <li class="culture"><a href="//news.daum.net/culture/" class="link_gnb link_gnb6"><span class="ir_wa">문화</span><span class="bar_gnb"><span class="inner_bar"></span></span></a></li> 
       <li class="digital"><a href="//news.daum.net/digital/" class="link_gnb link_gnb7"><span class="ir_wa">IT</span><span class="bar_gnb"><span class="inner_bar"></span></span></a></li> 
       <li class="series editorial"><a href="//news.daum.net/series" class="link_gnb link_gnb9"><span class="ir_wa">연재</span><span class="bar_gnb"><span class="inner_bar"></span></span></a></li> 
       <li class="photo"><a href="//news.daum.net/photo" class="link_gnb link_gnb10"><span class="ir_wa">포토</span><span class="bar_gnb"><span class="inner_bar"></span></span></a></li> 
       <li class="factcheck"> <a href="https://focus.daum.net/ch/news/factcheck" class="link_gnb link_gnb12"><span class="ir_wa">팩트체크</span><span class="bar_gnb"><span class="inner_bar"></span></span></a> </li> 
      </ul> 
      <div class="gnb_etc">  
       <div class="gnb_weather"> <strong class="screen_out">날씨 정보</strong> 
        <div id="gnb_weather"> 
         <ul class="list_weather" data-nightyn="Y" style="margin-top: 0px;"><li>
                          <a href="https://search.daum.net/search?nil_suggest=btn&amp;w=tot&amp;DA=SBC&amp;q=강릉%20날씨" class="link_cont">
                          <span class="ico_weather ico_weather1">강릉</span> 
                          <span class="txt_weather">강릉
                              <span class="num_heat">6</span> 
                              <span class="txt_heat">℃</span>
                          </span>
                          </a>
                        </li><li>
                          <a href="https://search.daum.net/search?nil_suggest=btn&amp;w=tot&amp;DA=SBC&amp;q=백령%20날씨" class="link_cont">
                          <span class="ico_weather ico_weather1">백령</span> 
                          <span class="txt_weather">백령
                              <span class="num_heat">5</span> 
                              <span class="txt_heat">℃</span>
                          </span>
                          </a>
                        </li><li>
                          <a href="https://search.daum.net/search?nil_suggest=btn&amp;w=tot&amp;DA=SBC&amp;q=청주%20날씨" class="link_cont">
                          <span class="ico_weather ico_weather1">청주</span> 
                          <span class="txt_weather">청주
                              <span class="num_heat">-1</span> 
                              <span class="txt_heat">℃</span>
                          </span>
                          </a>
                        </li><li>
                          <a href="https://search.daum.net/search?nil_suggest=btn&amp;w=tot&amp;DA=SBC&amp;q=전주%20날씨" class="link_cont">
                          <span class="ico_weather ico_weather1">전주</span> 
                          <span class="txt_weather">전주
                              <span class="num_heat">1</span> 
                              <span class="txt_heat">℃</span>
                          </span>
                          </a>
                        </li><li>
                          <a href="https://search.daum.net/search?nil_suggest=btn&amp;w=tot&amp;DA=SBC&amp;q=목포%20날씨" class="link_cont">
                          <span class="ico_weather ico_weather1">목포</span> 
                          <span class="txt_weather">목포
                              <span class="num_heat">2</span> 
                              <span class="txt_heat">℃</span>
                          </span>
                          </a>
                        </li><li>
                          <a href="https://search.daum.net/search?nil_suggest=btn&amp;w=tot&amp;DA=SBC&amp;q=여수%20날씨" class="link_cont">
                          <span class="ico_weather ico_weather1">여수</span> 
                          <span class="txt_weather">여수
                              <span class="num_heat">5</span> 
                              <span class="txt_heat">℃</span>
                          </span>
                          </a>
                        </li><li>
                          <a href="https://search.daum.net/search?nil_suggest=btn&amp;w=tot&amp;DA=SBC&amp;q=제주%20날씨" class="link_cont">
                          <span class="ico_weather ico_weather1">제주</span> 
                          <span class="txt_weather">제주
                              <span class="num_heat">3</span> 
                              <span class="txt_heat">℃</span>
                          </span>
                          </a>
                        </li><li>
                          <a href="https://search.daum.net/search?nil_suggest=btn&amp;w=tot&amp;DA=SBC&amp;q=안동%20날씨" class="link_cont">
                          <span class="ico_weather ico_weather1">안동</span> 
                          <span class="txt_weather">안동
                              <span class="num_heat">1</span> 
                              <span class="txt_heat">℃</span>
                          </span>
                          </a>
                        </li><li>
                          <a href="https://search.daum.net/search?nil_suggest=btn&amp;w=tot&amp;DA=SBC&amp;q=창원%20날씨" class="link_cont">
                          <span class="ico_weather ico_weather1">창원</span> 
                          <span class="txt_weather">창원
                              <span class="num_heat">3</span> 
                              <span class="txt_heat">℃</span>
                          </span>
                          </a>
                        </li><li>
                          <a href="https://search.daum.net/search?nil_suggest=btn&amp;w=tot&amp;DA=SBC&amp;q=서울%20날씨" class="link_cont">
                          <span class="ico_weather ico_weather1">서울</span> 
                          <span class="txt_weather">서울
                              <span class="num_heat">3</span> 
                              <span class="txt_heat">℃</span>
                          </span>
                          </a>
                        </li><li>
                          <a href="https://search.daum.net/search?nil_suggest=btn&amp;w=tot&amp;DA=SBC&amp;q=수원%20날씨" class="link_cont">
                          <span class="ico_weather ico_weather1">수원</span> 
                          <span class="txt_weather">수원
                              <span class="num_heat">1</span> 
                              <span class="txt_heat">℃</span>
                          </span>
                          </a>
                        </li><li>
                          <a href="https://search.daum.net/search?nil_suggest=btn&amp;w=tot&amp;DA=SBC&amp;q=인천%20날씨" class="link_cont">
                          <span class="ico_weather ico_weather1">인천</span> 
                          <span class="txt_weather">인천
                              <span class="num_heat">0</span> 
                              <span class="txt_heat">℃</span>
                          </span>
                          </a>
                        </li><li>
                          <a href="https://search.daum.net/search?nil_suggest=btn&amp;w=tot&amp;DA=SBC&amp;q=대구%20날씨" class="link_cont">
                          <span class="ico_weather ico_weather1">대구</span> 
                          <span class="txt_weather">대구
                              <span class="num_heat">3</span> 
                              <span class="txt_heat">℃</span>
                          </span>
                          </a>
                        </li><li>
                          <a href="https://search.daum.net/search?nil_suggest=btn&amp;w=tot&amp;DA=SBC&amp;q=대전%20날씨" class="link_cont">
                          <span class="ico_weather ico_weather1">대전</span> 
                          <span class="txt_weather">대전
                              <span class="num_heat">3</span> 
                              <span class="txt_heat">℃</span>
                          </span>
                          </a>
                        </li><li>
                          <a href="https://search.daum.net/search?nil_suggest=btn&amp;w=tot&amp;DA=SBC&amp;q=광주%20날씨" class="link_cont">
                          <span class="ico_weather ico_weather1">광주</span> 
                          <span class="txt_weather">광주
                              <span class="num_heat">0</span> 
                              <span class="txt_heat">℃</span>
                          </span>
                          </a>
                        </li><li>
                          <a href="https://search.daum.net/search?nil_suggest=btn&amp;w=tot&amp;DA=SBC&amp;q=부산%20날씨" class="link_cont">
                          <span class="ico_weather ico_weather1">부산</span> 
                          <span class="txt_weather">부산
                              <span class="num_heat">3</span> 
                              <span class="txt_heat">℃</span>
                          </span>
                          </a>
                        </li><li>
                          <a href="https://search.daum.net/search?nil_suggest=btn&amp;w=tot&amp;DA=SBC&amp;q=울산%20날씨" class="link_cont">
                          <span class="ico_weather ico_weather1">울산</span> 
                          <span class="txt_weather">울산
                              <span class="num_heat">4</span> 
                              <span class="txt_heat">℃</span>
                          </span>
                          </a>
                        </li><li>
                          <a href="https://search.daum.net/search?nil_suggest=btn&amp;w=tot&amp;DA=SBC&amp;q=울릉/독도%20날씨" class="link_cont">
                          <span class="ico_weather ico_weather1">울릉/독도</span> 
                          <span class="txt_weather">울릉/독도
                              <span class="num_heat">6</span> 
                              <span class="txt_heat">℃</span>
                          </span>
                          </a>
                        </li><li>
                          <a href="https://search.daum.net/search?nil_suggest=btn&amp;w=tot&amp;DA=SBC&amp;q=춘천%20날씨" class="link_cont">
                          <span class="ico_weather ico_weather1">춘천</span> 
                          <span class="txt_weather">춘천
                              <span class="num_heat">0</span> 
                              <span class="txt_heat">℃</span>
                          </span>
                          </a>
                        </li></ul> 
        </div> 
       </div> 
      </div> 
     </div> 
    </div> 
    <div id="barProgress" class="bar_progress" data-cloud="pc_header_gnb"></div>          
   </div> 
   <div id="dummyKakaoGnb" style="height:137px; display:none;"></div> 
   <hr class="hide"> 
   <div id="kakaoContent" class="cont_view" role="main"> 
    <h2 id="kakaoBody" class="screen_out">경제</h2> 
    <div id="cSub" data-cloud-area="title"> 
     <div class="head_view" data-cloud="pc_article_title_and_util"> <em class="info_cp"> <a href="http://www.khan.co.kr" class="link_cp" target="_blank"> <img src="https://t1.daumcdn.net/media/news/news2016/cp/cp_khan.gif" class="thumb_g" alt="경향신문"> </a> </em> 
      <h3 class="tit_view" data-translation="true">국제유가, 두 달 만에 30% 치솟아..정유·화학·항공·해운사 '초긴장'</h3> <span class="info_view"> <span class="txt_info">노정연 기자</span> <span class="txt_info">입력 <span class="num_date">2022. 02. 27. 22:49</span></span> <span class="txt_info">수정 <span class="num_date">2022. 02. 27. 23:57</span></span> <button id="alexCounter" class="emph_g2 link_cmt" style="cursor: pointer;">댓글 <span class="alex-count-area">2</span>개</button> </span> 
      <div class="util_view"> 
       <h4 class="screen_out">기사 도구 모음</h4>    
       <div class="btn_util translate_btn"> <button type="button" class="btn_set btn_translate #util #translate #set" aria-expanded="false"><span class="ico_newsview">번역 설정</span></button> 
        <div class="wrap_translatelayer"> 
         <div class="layer_util layer_translate"> <strong class="tit_translate">번역<span class="ico_newsview">beta</span> </strong> 
          <p class="desc_translate txt_newsview">Translated by <a href="https://kakao.ai/" target="_blank" class="link_kakaoi #util #translate #kakaoi_link">kakao i</a></p> <strong class="screen_out">번역할 언어 선택</strong> 
          <ul class="list_lang"> 
           <li> <button class="btn_lang btn_en #util #translate #lang_en" name="en"> <span class="ico_newsview ico_flag"></span> <strong class="ico_newsview tit_lang">English</strong> </button> </li> 
           <li> <button type="button" class="btn_lang btn_ch #util #translate #lang_cn" name="cn"> <span class="ico_newsview ico_flag"></span> <strong class="ico_newsview tit_lang">简体中文</strong> </button> </li> 
           <li> <button type="button" class="btn_lang btn_jp #util #translate #lang_jp" name="jp"> <span class="ico_newsview ico_flag"></span> <strong class="ico_newsview tit_lang">日本語</strong> </button> </li> 
           <li> <button type="button" class="btn_lang btn_ind #util #translate #lang_id" name="id"> <span class="ico_newsview ico_flag"></span> <strong class="ico_newsview tit_lang">Bahasa Indonesia</strong> </button> </li> 
           <li> <button type="button" class="btn_lang btn_viet #util #translate #lang_vi" name="vi"> <span class="ico_newsview ico_flag"></span> <strong class="ico_newsview tit_lang">tiếng Việt</strong> </button> </li> 
           <li> <button type="button" class="btn_lang btn_ko #util #translate #lang_kr" name="kr"> <span class="ico_newsview ico_flag"></span> <strong class="ico_newsview tit_lang">한국어</strong> </button> </li> 
          </ul> 
         </div> 
        </div> 
       </div>  
       <div class="btn_util share_btn"> <button id="rSocialShareButton" class="btn_set btn_share" aria-expanded="false"> <span class="ico_newsview">공유</span> </button> 
        <div id="rSocialShareLayerContainer"><div class="social_comm_p hide"><div class="sns_comm_p share_layer">
    <strong class="screen_out">SNS로 공유하기 펼쳐짐</strong>
    <ul class="list_sns"><li>
        <a href="#none" class="link_sns" data-handler="kakaotalk">
        <span class="img_social ico_sns_kt">
        </span>카카오톡</a></li><li>
        <a href="#none" class="link_sns" data-handler="kakaostory">
        <span class="img_social ico_sns_ks">
        </span>카카오스토리</a></li><li>
        <a href="#none" class="link_sns" data-handler="facebook">
        <span class="img_social ico_sns_fb">
        </span>페이스북</a></li><li>
        <a href="#none" class="link_sns" data-handler="twitter">
        <span class="img_social ico_sns_tw">
        </span>트위터</a></li><li>
        <a href="#none" class="link_sns" data-handler="mail">
        <span class="img_social ico_sns_mail">
        </span>다음메일</a></li><li>
        <a href="#none" class="link_sns" data-handler="daumcafe">
        <span class="img_social ico_sns_cafe">
        </span>다음카페</a></li></ul>
    <div class="copyUrl"><div class="sns_copyurl">
    <a href="#none" class="link_copyurl">
    <span class="screen_out">현재페이지 URL복사</span>
    <span class="txt_url">https://news.v.daum.net/v/20220227224906021</span>
    <span class="txt_copy">URL복사</span>
    </a>
    </div></div>
    <button class="btn_close"><span class="img_social ico_close">공유목록 닫기</span></button>
    </div><div class="sns_comm_p copyurl_layer hide">
  <div class="inner_copyurl">
  <p class="desc_copyurl">주소가 복사되었습니다.<br>원하는곳에 붙여넣기(Ctrl+V)해주세요.</p>
  <button class="btn_close"><span class="img_social ico_close">공유목록 닫기</span>
  </button>
  </div></div></div></div>  
       </div>  
       <div class="btn_util font_btn"> <button type="button" class="btn_set btn_font" aria-expanded="false"> <span class="ico_newsview ico_font">글씨크기 조절하기</span> </button> 
        <div class="wrap_setlayer"> 
         <div class="layer_util layer_set layer_font"> <strong class="screen_out">글자 크기 조절 레이어</strong> 
          <ul class="list_set list_font"> 
           <li> <input type="radio" id="inpFont17" name="inpFont" class="screen_out" value="17"> <label class="lab_set lab_size0 on" for="inpFont17"> <span class="ico_newsview"></span> <span aria-hidden="true">가나다라마바사</span> <span class="screen_out">매우 작은 폰트</span> </label> </li> 
           <li> <input type="radio" id="inpFont18" name="inpFont" class="screen_out" value="18"> <label class="lab_set lab_size1" for="inpFont18"> <span class="ico_newsview"></span> <span aria-hidden="true">가나다라마바사</span> <span class="screen_out">작은 폰트</span> </label> </li> 
           <li> <input type="radio" id="inpFont19" name="inpFont" class="screen_out" value="19"> <label class="lab_set lab_size2" for="inpFont19"> <span class="ico_newsview"></span> <span aria-hidden="true">가나다라마바사</span> <span class="screen_out">보통 폰트</span> </label> </li> 
           <li> <input type="radio" id="inpFont20" name="inpFont" class="screen_out" value="20"> <label class="lab_set lab_size3" for="inpFont20"> <span class="ico_newsview"></span> <span aria-hidden="true">가나다라마바사</span> <span class="screen_out">큰 폰트</span> </label> </li> 
           <li> <input type="radio" id="inpFont21" name="inpFont" class="screen_out" value="21"> <label class="lab_set lab_size4" for="inpFont21"> <span class="ico_newsview"></span> <span aria-hidden="true">가나다라마바사</span> <span class="screen_out">매우 큰 폰트</span> </label> </li> 
          </ul> 
         </div> 
        </div> 
       </div>  
       <div class="btn_util"> <button type="button" class="btn_print"><span class="ico_newsview">인쇄하기 새창열림</span></button> 
       </div> 
      </div> 
     </div>             
    </div> 
    <div id="cMain"> 
     <div id="mArticle"> 
      <div data-cloud-area="article">   
       <div id="translateNotice" class="status_view" style="display:none;" data-cloud="pc_article_body"> <span class="inner_status"> <p class="txt_status notice_lang notice_lang_en" style="display:none;">The copyright belongs to the original writer of the content, and there may be errors in machine translation results.</p> <p class="txt_status notice_lang notice_lang_cn" style="display:none;">版权归内容原作者所有。机器翻译结果可能存在错误。</p> <p class="txt_status notice_lang notice_lang_jp" style="display:none;">原文の著作権は原著著作者にあり、機械翻訳の結果にエラーが含まれることがあります。</p> <p class="txt_status notice_lang notice_lang_id" style="display:none;">Hak cipta milik penulis asli dari konten, dan mungkin ditemukan kesalahan dalam hasil terjemahan mesin.</p> <p class="txt_status notice_lang notice_lang_vi" style="display:none;">Bản quyền thuộc về tác giả gốc của nội dung và có thể có lỗi trong kết quả dịch bằng máy.</p> </span> 
       </div>  
       <div class="news_view" style="font-size:17px" data-cloud="pc_article_body"> <strong class="summary_view" data-translation="true"> 장기화로 비축분 소진 땐 대형 악재 </strong> 
        <div id="harmonyContainer" class="article_view" data-translation-body="true"> 
         <section dmcf-sid="EMTAdt9KHV">
          <p dmcf-pid="E53X0YniC8" dmcf-ptype="general">[경향신문] </p>
          <figure class="figure_frm origin_fig" dmcf-pid="EVpycypKDa" dmcf-ptype="figure">
           <p class="link_figure"><img alt="휘발유 가격 6주 연속 상승 27일 서울 시내 주유소에 기름값이 표시돼 있다. 한국석유공사에 따르면 2월 넷째주 전국 주유소의 평균 휘발유 가격은 ℓ당 1739.8원으로 전주보다 21.4원 오르며 6주 연속 상승세를 이어갔다. 연합뉴스" class="thumb_g_article" data-org-src="https://t1.daumcdn.net/news/202202/27/khan/20220227235701460hzla.jpg" data-org-width="400" dmcf-mid="ETjEXfgIOY" dmcf-mtype="image" height="auto" src="https://img1.daumcdn.net/thumb/R658x0.q70/?fname=https://t1.daumcdn.net/news/202202/27/khan/20220227235701460hzla.jpg" width="658"></p>
           <figcaption class="txt_caption default_figure">
            휘발유 가격 6주 연속 상승 27일 서울 시내 주유소에 기름값이 표시돼 있다. 한국석유공사에 따르면 2월 넷째주 전국 주유소의 평균 휘발유 가격은 ℓ당 1739.8원으로 전주보다 21.4원 오르며 6주 연속 상승세를 이어갔다. 연합뉴스
           </figcaption>
          </figure>
          <p dmcf-pid="EsvITxn8nw" dmcf-ptype="general">러시아의 우크라이나 침공으로 국제유가 고공행진이 계속되며 산업계 우려가 커지고 있다. 우크라이나 사태가 장기화될 경우 원유를 직접 수입해 판매·활용하는 정유·화학업계를 비롯해 항공과 해운업계에도 파장이 예상된다.</p>
          <p dmcf-pid="Ea2Ka5W6tc" dmcf-ptype="general">27일 한국석유공사 유가정보사이트 오피넷에 따르면 두바이유 가격은 2월 넷째주 평균 95.0달러로 집계됐다. 전주보다 3.1% 상승했으며 지난해 12월 평균 가격과 비교하면 약 두 달 만에 30% 가까이 뛰었다. 미국 정부가 유가 급등으로 인한 피해를 고려해 러시아산 원유를 제재 대상에서 제외하긴 했으나 시장에서는 전쟁의 영향으로 유가의 고공행진이 불가피하다는 전망이 많다.</p>
          <p dmcf-pid="EPPncxYckV" dmcf-ptype="general">원유를 수입해 국내에서 정제해 판매하는 한국 정유사들은 보통 유가 급등의 수혜자로 거론된다. 정유사들은 일정량의 원유를 비축해두는데, 유가가 오르면 비축해둔 원유의 가치가 상승하면서 ‘재고 평가이익’이 커지기 때문이다. 하지만 고유가가 장기화할 경우 석유제품 인상으로 수요가 위축될 가능성이 크고 이는 곧 수익성 악화로 이어질 수 있다.</p>
          <p dmcf-pid="ERldDTzmg6" dmcf-ptype="general">석유화학업계는 유가 강세로 나프타 가격도 함께 올라 원가 부담이 커지고 있다. 석유화학업체들이 사용하는 나프타 중 외국산 비중은 약 20%이며 외국산 가운데 러시아산은 약 23%다. 러시아산 나프타 수입이 제한되면 기업들의 공급망에 위협이 될 뿐 아니라 추가 가격 상승으로 이어질 수 있다. </p>
          <p dmcf-pid="EVqGVmJidh" dmcf-ptype="general">석유화학업계 관계자는 “재고 비축분이 있어 당장 직접적인 피해는 없을 것”이라면서도 “차질이 생길 가능성에 대비해 중동과 남미 등으로 공급처를 늘리는 방안을 검토하고 있다”고 말했다.</p>
          <p dmcf-pid="E4fXXsYsGj" dmcf-ptype="general">코로나19 이후 화물 운송을 늘리며 수익을 내고 있는 항공업계는 유가 상승으로 상황이 반전될 가능성이 크다. 항공사는 전체 영업비용에서 유류비가 약 25~30%를 차지한다. 대한항공은 유가가 배럴당 1달러 오르면 연간 약 3000만달러(약 361억원)의 손실이 발생할 것으로 추산하고 있다. 해운업계 역시 컨테이너선 운용 비용 중 20%가량을 연료비에 사용한다.</p>
          <p dmcf-pid="ECanMxoeaj" dmcf-ptype="general">노정연 기자 dana_fm@kyunghyang.com</p>
         </section> 
         <p data-translation="true">ⓒ 경향신문 &amp; 경향닷컴(www.khan.co.kr), 무단전재 및 재배포 금지 </p> 
        </div> 
       </div>  
      </div> 
      <div class="foot_view" data-cloud-area="addition"> 
       <div class="emotion_wrap" data-cloud="alex_action_emotion"> 
        <div class="emotion_tit"> <strong class="tit_emotion">이 기사에 대해 어떻게 생각하시나요?</strong> 
        </div> 
        <div class="emotion_list"> 
         <div id="alex_action_emotion" class="alex-actions" data-client-id="26BXAvKny5WF5Z09lr5k77Y8" data-item-key="@20220227224906021" data-action-mode="REACTION" data-message-channel="alex-action-reactions" data-selections="[{&quot;type&quot;:&quot;RECOMMEND&quot;,&quot;animation&quot;:&quot;SLIDE&quot;,&quot;image&quot;:&quot;https://t1.daumcdn.net/harmony_static/alex_emotion/SLIDE_RECOMMEND_1x.png&quot;,&quot;highResImage&quot;:&quot;https://t1.daumcdn.net/harmony_static/alex_emotion/SLIDE_RECOMMEND_2x.png&quot;},{&quot;type&quot;:&quot;LIKE&quot;,&quot;animation&quot;:&quot;SLIDE&quot;,&quot;image&quot;:&quot;https://t1.daumcdn.net/harmony_static/alex_emotion/SLIDE_LIKE_1x.png&quot;,&quot;highResImage&quot;:&quot;https://t1.daumcdn.net/harmony_static/alex_emotion/SLIDE_LIKE_2x.png&quot;},{&quot;type&quot;:&quot;IMPRESS&quot;,&quot;animation&quot;:&quot;SLIDE&quot;,&quot;image&quot;:&quot;https://t1.daumcdn.net/harmony_static/alex_emotion/SLIDE_IMPRESS_1x.png&quot;,&quot;highResImage&quot;:&quot;https://t1.daumcdn.net/harmony_static/alex_emotion/SLIDE_IMPRESS_2x.png&quot;},{&quot;type&quot;:&quot;ANGRY&quot;,&quot;animation&quot;:&quot;SLIDE&quot;,&quot;image&quot;:&quot;https://t1.daumcdn.net/harmony_static/alex_emotion/SLIDE_ANGRY_1x.png&quot;,&quot;highResImage&quot;:&quot;https://t1.daumcdn.net/harmony_static/alex_emotion/SLIDE_ANGRY_2x.png&quot;},{&quot;type&quot;:&quot;SAD&quot;,&quot;animation&quot;:&quot;SLIDE&quot;,&quot;image&quot;:&quot;https://t1.daumcdn.net/harmony_static/alex_emotion/SLIDE_SAD_1x.png&quot;,&quot;highResImage&quot;:&quot;https://t1.daumcdn.net/harmony_static/alex_emotion/SLIDE_SAD_2x.png&quot;}]"><div data-tiara-custom="alex.component=REACTION&amp;alex.clientId=26BXAvKny5WF5Z09lr5k77Y8&amp;alex.itemKey=20220227224906021&amp;alex.forumKey=news" class="jsx-3762130795 🎬_main"><div class="jsx-637502918 🎬_selections"><button data-action-type="RECOMMEND" data-tiara-action-name="액션_추천해요" data-tiara-custom="alex.topType=화나요" class="jsx-2157231875 🎬_selection 🎬_zero_count"><div data-animation="SLIDE" class="jsx-2717597395 🎬_icon"></div><div class="jsx-2157231875 🎬_selection_label">추천해요</div><span class="jsx-2157231875 🎬_count_label">0</span></button><button data-action-type="LIKE" data-tiara-action-name="액션_좋아요" data-tiara-custom="alex.topType=화나요" class="jsx-2157231875 🎬_selection 🎬_zero_count"><div data-animation="SLIDE" class="jsx-3429346540 🎬_icon"></div><div class="jsx-2157231875 🎬_selection_label">좋아요</div><span class="jsx-2157231875 🎬_count_label">0</span></button><button data-action-type="IMPRESS" data-tiara-action-name="액션_감동이에요" data-tiara-custom="alex.topType=화나요" class="jsx-2157231875 🎬_selection 🎬_zero_count"><div data-animation="SLIDE" class="jsx-1617167524 🎬_icon"></div><div class="jsx-2157231875 🎬_selection_label">감동이에요</div><span class="jsx-2157231875 🎬_count_label">0</span></button><button data-action-type="ANGRY" data-tiara-action-name="액션_화나요" data-tiara-custom="alex.topType=화나요" class="jsx-2157231875 🎬_selection"><div data-animation="SLIDE" class="jsx-1027214948 🎬_icon"></div><div class="jsx-2157231875 🎬_selection_label">화나요</div><span class="jsx-2157231875 🎬_count_label">1</span></button><button data-action-type="SAD" data-tiara-action-name="액션_슬퍼요" data-tiara-custom="alex.topType=화나요" class="jsx-2157231875 🎬_selection"><div data-animation="SLIDE" class="jsx-2084876625 🎬_icon"></div><div class="jsx-2157231875 🎬_selection_label">슬퍼요</div><span class="jsx-2157231875 🎬_count_label">1</span></button></div></div></div> 
        </div> 
       </div>   
          
       <div class="cp_view" data-cloud="pc_related_news"> 
        <div class="cp_tit"> 
         <h3 class="tit_cp">경향신문 주요 뉴스</h3> 
         <p class="info_cp">해당 언론사로 연결됩니다.</p> 
        </div> 
        <ul class="list_relate"> 
         <li><a href="https://www.khan.co.kr/politics/assembly/article/202202271132001/?utm_campaign=daum_news&amp;utm_source=daum&amp;utm_medium=related_news" title="송영길 “대선 결과 어떻든, 민주당은 172석으로 대장동 특검 수용케할 것”" class="link_txt" target="_blank" data-type="rss">송영길 “대선 결과 어떻든, 민주당은 172석으로 대장동 특검 수용케할 것”</a></li> 
         <li><a href="https://www.khan.co.kr/politics/assembly/article/202202271621001/?utm_campaign=daum_news&amp;utm_source=daum&amp;utm_medium=related_news" title="안철수 “윤석열 측 신뢰 어려운 세력…립서비스만 계속”" class="link_txt" target="_blank" data-type="rss">안철수 “윤석열 측 신뢰 어려운 세력…립서비스만 계속”</a></li> 
         <li><a href="http://sports.khan.co.kr/news/sk_index.html?art_id=202202271426003&amp;code=540301&amp;utm_campaign=daum_news&amp;utm_source=daum&amp;utm_medium=related_news" title="“모두 빨리 코로나 걸렸으면” 양준일 망언 일파만파" class="link_txt" target="_blank" data-type="rss">“모두 빨리 코로나 걸렸으면” 양준일 망언 일파만파</a></li> 
         <li><a href="https://www.khan.co.kr/politics/assembly/article/202202240839001/?utm_campaign=daum_news&amp;utm_source=daum&amp;utm_medium=related_news" title="이준석, 이재명 '겁대가리' 발언에 &quot;이성 잃어가, 극좌 포퓰리스트 되려하나&quot;" class="link_txt" target="_blank" data-type="rss">이준석, 이재명 '겁대가리' 발언에 "이성 잃어가, 극좌 포퓰리스트 되려하나"</a></li> 
         <li><a href="https://www.khan.co.kr/world/world-general/article/202202272229001/?utm_campaign=daum_news&amp;utm_source=daum&amp;utm_medium=related_news" title="푸틴, 군 지휘부에 '핵 억제력' 강화 지시" class="link_txt" target="_blank" data-type="rss">푸틴, 군 지휘부에 '핵 억제력' 강화 지시</a></li> 
         <li><a href="https://www.khan.co.kr/national/incident/article/202202221928001/?utm_campaign=daum_news&amp;utm_source=daum&amp;utm_medium=related_news" title="국민의힘 의원 비서, ATM에 놓인 현금 가로채 달아나···경찰 &quot;추적중&quot;" class="link_txt" target="_blank" data-type="rss">국민의힘 의원 비서, ATM에 놓인 현금 가로채 달아나···경찰 "추적중"</a></li> 
         <li><a href="https://www.khan.co.kr/politics/election/article/202202241101001/?utm_campaign=daum_news&amp;utm_source=daum&amp;utm_medium=related_news" title="민주당 “김건희, 윤석열 서울중앙지검장 때도 작전주 투자 의혹”" class="link_txt" target="_blank" data-type="rss">민주당 “김건희, 윤석열 서울중앙지검장 때도 작전주 투자 의혹”</a></li> 
         <li><a href="https://www.khan.co.kr/environment/environment-general/article/202201171730001/?utm_campaign=daum_news&amp;utm_source=daum&amp;utm_medium=related_news" title="“제발 그만해, 이러다 다 죽어”···‘60+’ 할매·할배 '기후 행동' 뭉쳤다" class="link_txt" target="_blank" data-type="rss">“제발 그만해, 이러다 다 죽어”···‘60+’ 할매·할배 '기후 행동' 뭉쳤다</a></li> 
         <li><a href="https://www.khan.co.kr/politics/defense-diplomacy/article/202202140639001/?utm_campaign=daum_news&amp;utm_source=daum&amp;utm_medium=related_news" title="'병사 머리’는 보이고 '장교·부사관 벗은 몸’은 안 보이는 서욱 국방" class="link_txt" target="_blank" data-type="rss">'병사 머리’는 보이고 '장교·부사관 벗은 몸’은 안 보이는 서욱 국방</a></li> 
         <li><a href="https://www.khan.co.kr/economy/economy-general/article/202201301128011/?utm_campaign=daum_news&amp;utm_source=daum&amp;utm_medium=related_news" title="10억 들여 키운 '국가대표 씨수소' 1마리 가치 2940억원...&quot;자식만 7만마리'&quot;" class="link_txt" target="_blank" data-type="rss">10억 들여 키운 '국가대표 씨수소' 1마리 가치 2940억원..."자식만 7만마리'"</a></li> 
        </ul> 
       </div> 
       <div id="alex-scroll-area" data-cloud="pc_article_bottom_ad"></div> 
       <ins class="kakao_ad_area recomm_ad" style="display: block; text-decoration: none; height: 120px; max-height: inherit; min-height: 120px; min-width: 655px; width: 655px;" data-ad-unit="DAN-28ewRUAQYNwThl1H" data-ad-width="655" data-ad-height="120" data-ad-onfail="daRenderFailCallBack" data-cloud="pc_article_bottom_ad" data-ad-param-channel="harmony" data-ad-param-cp="11_pc_media_news" id="kakao_ad_ZcNb1i"><div data-ad-creative-wrap="outer" class="adfit__html-banner" style="min-width: 655px; min-height: 120px;"><div data-ad-creative-wrap="inner" class="adfit__html-banner__inner" style="height: 120px;"><div><iframe src="https://t1.daumcdn.net/adfit/adunit_style/5d41b91287f90788f7ee6d107c80fac1f8b3a750" width="100%" style="position: absolute; top: 0px; left: 0px; display: block; border: none; height: 120px; width: 655px; min-width: 655px; min-height: 120px; overflow: hidden; padding: 0px; margin: 0px auto;" frameborder="0" scrolling="no" marginheight="0" marginwidth="0" name="%5B%7B%22usePrefersColorScheme%22%3A%22N%22%2C%22usePreferColorScheme%22%3A%22N%22%2C%22frameId%22%3A%2296083f5a-4bf4-47f5-9342-0809b9a7e2b7%22%2C%22frameName%22%3A%2296083f5a-4bf4-47f5-9342-0809b9a7e2b7%22%2C%22title%22%3A%22%EC%89%BF!%20%ED%95%98%ED%95%84%20%EC%A7%80%EA%B8%88%20%ED%95%A0%EC%9D%B8%20%EC%A4%91%22%2C%22mainImage%22%3A%7B%22url%22%3A%22https%3A%2F%2Ft1.daumcdn.net%2Fb2%2Fcreative%2F338592%2F928f27ae85e06904d1fc8146ba151f76.jpg%22%2C%22width%22%3A1200%2C%22height%22%3A600%7D%2C%22profileName%22%3A%22%EC%8D%A8%EB%B8%8C%EC%9B%A8%EC%9D%B4%22%2C%22adInfoIcon%22%3A%7B%22url%22%3A%22https%3A%2F%2Ft1.daumcdn.net%2Fbiz%2Fui%2Fad%2FADmark%2Fi_mark_200803.png%22%2C%22width%22%3A15%2C%22height%22%3A15%7D%2C%22adInfoUrl%22%3A%22https%3A%2F%2Finfo.ad.daum.net%2Foptout.do%22%2C%22landingUrl%22%3A%22https%3A%2F%2Ftr.ad.daum.net%2Fclk%3Fwa%3DGk7MasF9HNdUIHPIYBLKIbl98uBgJRXCQYisQlsMkaA%26enc%3DBx1E0yYc9vEInEtrKjLMYEkm5crwC5t4PhFBGjeaOb3JIUVVUD0PZV3hkSxmDtBx1SGlUEYqoDRAyVcA_hRIoCl0-csPM4_tp0NjKYDgx3GjkzZFTptTIVzo4gllctwAVvfeZ-WtMxndDIt3qK4EX5ytQmgjGd5-5F7-aE67tuBCwa78FafIahqUHrAJllQLXebaOVoMPCEq4EvKmwkTDgeVYSUlJxcI4pLO8kliPBqUyTGaMUY7aDjW8KDfOtQb24go56Ur2UGPhBJnKiA5aWv3BveVV4Cb_Wj49xNxk17jhNuYUQWhYma28RQw8PKLLYa7JZLSnI8G64kkLiYe2nMkhu7LaESODKQGA5b1K2CkxtYrr6Tl4mWCczgo3aIOSZJx-732QfkH3T7pLTPVfVVqHTl8sNy_68XBKD8Y0IPuEOhiJLoRs-8pl9nCP-5t%22%2C%22useAdInfoIcon%22%3Afalse%2C%22useAdInfoUrl%22%3Afalse%2C%22events%22%3A%5B%7B%22type%22%3A%22viewable%22%2C%22url%22%3A%22https%3A%2F%2Ftr.ad.daum.net%2Fvimp%3Fwa%3DGk7MasF9HNdUIHPIYBLKIbl98uBgJRXCQYisQlsMkaA%26enc%3DBx1E0yYc9vEInEtrKjLMYEkm5crwC5t4PhFBGjeaOb3JIUVVUD0PZV3hkSxmDtBx1SGlUEYqoDRAyVcA_hRIoCl0-csPM4_tp0NjKYDgx3GjkzZFTptTIVzo4gllctwAVvfeZ-WtMxndDIt3qK4EX5ytQmgjGd5-5F7-aE67tuBCwa78FafIahqUHrAJllQLXebaOVoMPCEq4EvKmwkTDgeVYSUlJxcI4pLO8kliPBqUyTGaMUY7aDjW8KDfOtQb24go56Ur2UGPhBJnKiA5aUdOsvvORPB4LQ8z50jX6O3lecQYrgfcUwJwYd_5hHDdxbGT94OmQdoqmN0fF2OUMVlaATiO9KsjEnU2GbDSz1I%22%7D%2C%7B%22type%22%3A%22hide%22%2C%22url%22%3A%22https%3A%2F%2Ftr.ad.daum.net%2Fac%3Fwa%3DGk7MasF9HNdUIHPIYBLKIbl98uBgJRXCQYisQlsMkaA%26enc%3DBx1E0yYc9vEInEtrKjLMYEkm5crwC5t4PhFBGjeaOb3JIUVVUD0PZV3hkSxmDtBx1SGlUEYqoDRAyVcA_hRIoCl0-csPM4_tp0NjKYDgx3GjkzZFTptTIVzo4gllctwAVvfeZ-WtMxndDIt3qK4EX5ytQmgjGd5-5F7-aE67tuBCwa78FafIahqUHrAJllQLXebaOVoMPCEq4EvKmwkTDgeVYSUlJxcI4pLO8kliPBqUyTGaMUY7aDjW8KDfOtQb24go56Ur2UGPhBJnKiA5aUdOsvvORPB4LQ8z50jX6O3lecQYrgfcUwJwYd_5hHDdxbGT94OmQdoqmN0fF2OUMYVzvex5F1KQoSJ_ftowwU8%22%7D%2C%7B%22type%22%3A%22rendered%22%2C%22url%22%3A%22https%3A%2F%2Fkyson.ad.daum.net%2Fad%2Frendered%3Frequest_id%3D9b43eb67-edbf-4111-a1d5-50198f0c5bdd%26adunit_id%3DDAN-28ewRUAQYNwThl1H%26bid_id%3D20220228021941_2apPj%26dsp_id%3DMOMENT%26med_dsp_id%3DMOMENT%26network_type%3DDAUM%26sdk_type%3Dweb%26sdk_version%3D4.1.5%26ad_type%3DBanner%26w%3D655%26h%3D120%26template_seq%3D41%22%7D%2C%7B%22type%22%3A%22viewable%22%2C%22url%22%3A%22https%3A%2F%2Fkyson.ad.daum.net%2Fad%2Fviewable%3Frequest_id%3D9b43eb67-edbf-4111-a1d5-50198f0c5bdd%26adunit_id%3DDAN-28ewRUAQYNwThl1H%26bid_id%3D20220228021941_2apPj%26dsp_id%3DMOMENT%26med_dsp_id%3DMOMENT%26network_type%3DDAUM%26sdk_type%3Dweb%26sdk_version%3D4.1.5%26ad_type%3DBanner%26w%3D655%26h%3D120%26template_seq%3D41%22%7D%2C%7B%22type%22%3A%22click%22%2C%22url%22%3A%22https%3A%2F%2Fkyson.ad.daum.net%2Fad%2Fclick%3Frequest_id%3D9b43eb67-edbf-4111-a1d5-50198f0c5bdd%26adunit_id%3DDAN-28ewRUAQYNwThl1H%26bid_id%3D20220228021941_2apPj%26dsp_id%3DMOMENT%26med_dsp_id%3DMOMENT%26network_type%3DDAUM%26sdk_type%3Dweb%26sdk_version%3D4.1.5%26ad_type%3DBanner%26w%3D655%26h%3D120%26template_seq%3D41%22%7D%2C%7B%22type%22%3A%22duration%22%2C%22url%22%3A%22https%3A%2F%2Fkyson.ad.daum.net%2Fad%2Fduration%3Frequest_id%3D9b43eb67-edbf-4111-a1d5-50198f0c5bdd%26adunit_id%3DDAN-28ewRUAQYNwThl1H%26bid_id%3D20220228021941_2apPj%26dsp_id%3DMOMENT%26med_dsp_id%3DMOMENT%26network_type%3DDAUM%26sdk_type%3Dweb%26sdk_version%3D4.1.5%26ad_type%3DBanner%26w%3D655%26h%3D120%26template_seq%3D41%22%7D%5D%2C%22dspId%22%3A%22MOMENT%22%2C%22altText%22%3A%22%EC%8D%A8%EB%B8%8C%EC%9B%A8%EC%9D%B4%20%EC%9D%B4%ED%83%88%EB%A6%AC%EC%95%88%20%EB%B9%84%EC%97%A0%ED%8B%B0%22%2C%22nasTemplateSeq%22%3A41%2C%22type%22%3A%22native%22%7D%5D" id="96083f5a-4bf4-47f5-9342-0809b9a7e2b7"></iframe></div></div><div class="adfit__border adfit__border--left" data-ad-creative-wrap="border" style="background-color: rgb(229, 229, 229);"></div><div class="adfit__border adfit__border--right" data-ad-creative-wrap="border" style="background-color: rgb(229, 229, 229);"></div><div class="adfit__border adfit__border--top" data-ad-creative-wrap="border" style="background-color: rgb(229, 229, 229);"></div><div class="adfit__border adfit__border--bottom" data-ad-creative-wrap="border" style="background-color: rgb(229, 229, 229);"></div></div> </ins>  
       <div class="cmt_news cmt_view" data-cloud="pc_comment" style="min-height: 250px;"> 
        <div id="newsviewNotice" class="news_notice"><strong class="tit_notice">공지</strong><span class="txt_notice"><a href="https://magazine.kakao.com/docs/620cab4483e94101ac3e4929" class="link_txt">제20대 대통령 선거 댓글 작성시 유의사항 안내</a></span></div>  
        <div id="alex-area" class="alex-area" data-base-url="//comment.daum.net" data-post-id="@20220227224906021" data-client-id="26BXAvKny5WF5Z09lr5k77Y8" data-kakao-appkey="9076ba452fea1a434141d6b6a2c11a3c" data-initial-limit="3" data-page-limit="10" data-sort="P" data-use-my="true" data-max-length="300" data-use-reply="true" data-use-official="true" data-use-emoticon="true" data-use-social-share="true" data-use-attachment="false" data-is-mobile="false" data-show-mask-alert="true" data-use-live-comment-mode="notify" data-use-fold="true" data-use-skin="newsView" data-use-safebot="true" data-use-force-fold="true"><div class="pc" data-reactid=".0"><div class="alex_single" data-tiara-custom="alex.component=single&amp;alex.clientId=26BXAvKny5WF5Z09lr5k77Y8&amp;alex.forumId=-99&amp;alex.postKey=20220227224906021" data-reactid=".0.0"><div class="wrap_alex #comment fold" data-tiara-custom="alex.view=main" data-reactid=".0.0.0"><noscript data-reactid=".0.0.0.0"></noscript><div class="cmt_count_box" data-reactid=".0.0.0.1"><span id="alex-header" class="cmt_count" data-reactid=".0.0.0.1.0"><span data-reactid=".0.0.0.1.0.0">댓글</span><span data-reactid=".0.0.0.1.0.1"> </span><em class="num_count" data-reactid=".0.0.0.1.0.2">2</em></span><button class="btn_foldup" aria-expanded="false" data-tiara-action-name="댓글영역_접기 취소" data-reactid=".0.0.0.1.1"><span class="img_cmt ico_fold" data-reactid=".0.0.0.1.1.0"><span class="screen_out" data-reactid=".0.0.0.1.1.0.0">댓글영역 접기 취소</span></span></button><a role="button" href="javascript:" class="my_comments #my" data-tiara-action-name="내댓글활동_보기" data-tiara-custom="alex.withNewNoti=false" data-reactid=".0.0.0.1.2"><span data-reactid=".0.0.0.1.2.0">MY</span></a><div class="tooltip_fold" data-reactid=".0.0.0.1.3"><a role="button" href="#" class="btn_close" data-reactid=".0.0.0.1.3.0"><span class="img_cmt" data-reactid=".0.0.0.1.3.0.0">닫기</span></a><div class="cont_fold" data-reactid=".0.0.0.1.3.1">댓글 접기가 적용되었습니다. 
해당 서비스에서는 댓글이 항상 접힌 상태로 보이게 됩니다. 
댓글을 보고 싶으면 버튼을 다시 클릭하면 됩니다.</div></div><div class="forcefold_guide on" data-reactid=".0.0.0.1.4">댓글 영역은 접힘 상태로 기본 제공되며, on/off 버튼을 통해 댓글을 확인하실 수 있습니다.</div></div><div class="reply_write_box" data-reactid=".0.0.0.2"><div id="writeForm_comment" class="write_wrap" data-reactid=".0.0.0.2.1"><fieldset data-reactid=".0.0.0.2.1.0"><legend class="screen_out" data-reactid=".0.0.0.2.1.0.0">댓글 입력 폼</legend><a role="button" href="javascript:" id="alex_login_box_comment" class="link_write" data-reactid=".0.0.0.2.1.0.1"><span data-reactid=".0.0.0.2.1.0.1.0"><span class="img_cmt img_emot" data-reactid=".0.0.0.2.1.0.1.0.0"></span></span><span class="txt_write" data-reactid=".0.0.0.2.1.0.1.1">로그인 해주세요.</span></a><div class="cont_write" data-reactid=".0.0.0.2.1.0.2"><div class="write_alex" data-reactid=".0.0.0.2.1.0.2.0"><noscript data-reactid=".0.0.0.2.1.0.2.0.0"></noscript><div data-reactid=".0.0.0.2.1.0.2.0.1"><div id="alex_write_box" class="alex_write" data-reactid=".0.0.0.2.1.0.2.0.1.0"><div id="alex_input_wrap" class="input_wrap" data-reactid=".0.0.0.2.1.0.2.0.1.0.1"><label id="commentLabel" class="lab_cmt" for="commentContent" data-reactid=".0.0.0.2.1.0.2.0.1.0.1.0"><span class="txt_info" data-reactid=".0.0.0.2.1.0.2.0.1.0.1.0.1">투표 관련 허위사실을 유포하여 선거의 자유를 방해하거나, 후보자 등에 관한 허위사실･비방을 게시하는 경우에는 공직선거법에 따라 댓글이 삭제되거나 고발될 수 있습니다.</span></label><div class="alex_highlight_wrap" data-reactid=".0.0.0.2.1.0.2.0.1.0.1.1"></div><textarea title="댓글 내용 입력 공간" class="tf_cmt" id="commentContent" name="commentContent" data-reactid=".0.0.0.2.1.0.2.0.1.0.1.2"></textarea></div></div><div class="alex_console" data-reactid=".0.0.0.2.1.0.2.0.1.1"><div class="txt_byte" data-reactid=".0.0.0.2.1.0.2.0.1.1.0"><span class="num_fst" id="commentByte" data-reactid=".0.0.0.2.1.0.2.0.1.1.0.0">0</span><span class="slash" data-reactid=".0.0.0.2.1.0.2.0.1.1.0.1">/</span><span data-reactid=".0.0.0.2.1.0.2.0.1.1.0.2">300</span></div><button class="btn_emot" data-reactid=".0.0.0.2.1.0.2.0.1.1.1"><span class="img_cmt img_emot" data-reactid=".0.0.0.2.1.0.2.0.1.1.1.0">이모티콘 입력하기</span></button><p class="txt_guide" data-reactid=".0.0.0.2.1.0.2.0.1.1.3">※ 이 댓글에 대한 법적 책임은 작성자에게 귀속됩니다.</p></div><div class="tool_write" data-reactid=".0.0.0.2.1.0.2.0.1.2"><button class="btn_write write_type2 #write" disabled="" type="submit" data-tiara-action-kind="WriteComment" data-tiara-action-name="댓글_등록" data-reactid=".0.0.0.2.1.0.2.0.1.2.1"><span class="inner_btn" data-reactid=".0.0.0.2.1.0.2.0.1.2.1.0">등록</span></button></div><div class="emoticon_keyboard_wrapper" data-reactid=".0.0.0.2.1.0.2.0.1.3"><div id="commentEmoticonLayer" class="emoticon_layer alex_emot" style="display:none;" data-reactid=".0.0.0.2.1.0.2.0.1.3.0"></div></div></div></div></div></fieldset></div></div><div class="cmt_box" data-reactid=".0.0.0.3"><ul class="list_category" data-reactid=".0.0.0.3.0"><li class="" data-reactid=".0.0.0.3.0.$recommend"><button class="link_cate #recommend" data-alex-sort="RECOMMEND" data-tiara-action-name="댓글목록_정렬 찬반순" data-reactid=".0.0.0.3.0.$recommend.0"><span data-reactid=".0.0.0.3.0.$recommend.0.0"><span data-reactid=".0.0.0.3.0.$recommend.0.0.0">찬반순</span></span></button><span data-reactid=".0.0.0.3.0.$recommend.1"></span></li><li class="on" data-reactid=".0.0.0.3.0.$recent"><button class="link_cate #latest" data-alex-sort="LATEST" data-tiara-action-name="댓글목록_정렬 최신순" data-reactid=".0.0.0.3.0.$recent.0"><span data-reactid=".0.0.0.3.0.$recent.0.0"><span data-reactid=".0.0.0.3.0.$recent.0.0.0">최신순</span><span class="screen_out" data-reactid=".0.0.0.3.0.$recent.0.0.1">선택됨</span></span></button><span data-reactid=".0.0.0.3.0.$recent.1"></span></li><li class="" data-reactid=".0.0.0.3.0.$old"><button class="link_cate #chronological" data-alex-sort="CHRONOLOGICAL" data-tiara-action-name="댓글목록_정렬 과거순" data-reactid=".0.0.0.3.0.$old.0"><span data-reactid=".0.0.0.3.0.$old.0.0"><span data-reactid=".0.0.0.3.0.$old.0.0.0">과거순</span></span></button><span data-reactid=".0.0.0.3.0.$old.1"></span></li></ul><div class="cmt_keeper_setting" data-reactid=".0.0.0.3.1"><div class="tooltip_box_1" data-reactid=".0.0.0.3.1.0">세이프봇 설정</div><button class="btn_setting" data-tiara-action-name="세이프봇_설정 열기" data-reactid=".0.0.0.3.1.1"><span class="img_cmt_v1 ico_setting" data-reactid=".0.0.0.3.1.1.0">설정</span></button></div><button class="alex_btn btn_refresh #refresh" data-tiara-action-name="댓글목록_새로고침" data-reactid=".0.0.0.3.2"><span class="img_cmt ico_refresh" data-reactid=".0.0.0.3.2.0">새로고침</span></button><div class="new_cmt_container" data-reactid=".0.0.0.3.3"><div class="new_cmt_btn error_banner hide" aria-hidden="true" data-reactid=".0.0.0.3.3.0"><div class="banner_ico" data-reactid=".0.0.0.3.3.0.0"></div><span class="banner_msg" data-reactid=".0.0.0.3.3.0.1">댓글 실시간 업데이트가 지연되고 있습니다.</span></div><div class="new_cmt_btn hide" aria-hidden="true" data-reactid=".0.0.0.3.3.1"><span data-reactid=".0.0.0.3.3.1.0">새로운 댓글</span><span data-reactid=".0.0.0.3.3.1.1"> </span><span class="new_cmt_count" data-reactid=".0.0.0.3.3.1.2">0</span></div></div><ul class="list_comment" data-reactid=".0.0.0.3.4"><li id="comment763650660" class="fst" data-reactid=".0.0.0.3.4.$763650660"><div class="cmt_info" data-reactid=".0.0.0.3.4.$763650660.0"><strong class="tit_nick" data-reactid=".0.0.0.3.4.$763650660.0.0"><noscript data-reactid=".0.0.0.3.4.$763650660.0.0.0"></noscript><span class="info_author" data-reactid=".0.0.0.3.4.$763650660.0.0.1"><span class="link_nick" data-reactid=".0.0.0.3.4.$763650660.0.0.1.0"><span class="img_cmt img_daum" data-reactid=".0.0.0.3.4.$763650660.0.0.1.0.0"></span></span><a role="button" href="javascript:" class="link_nick" data-userid="-11381000" data-tiara-action-name="다른사람댓글활동_보기" data-reactid=".0.0.0.3.4.$763650660.0.0.1.1"><span class="screen_out" data-reactid=".0.0.0.3.4.$763650660.0.0.1.1.0">댓글 작성자</span><span data-reactid=".0.0.0.3.4.$763650660.0.0.1.1.1">LUV_XRP</span><span class="screen_out" data-reactid=".0.0.0.3.4.$763650660.0.0.1.1.2">, 댓글모아보기</span></a><span class="txt_date" data-reactid=".0.0.0.3.4.$763650660.0.0.1.2">1시간전</span></span></strong><p class="desc_txt font_size_17" data-reactid=".0.0.0.3.4.$763650660.0.1:$desc-763650660">올릴때는 기존 사놓은 재고  소진안하고 바로 적용ㅋㅋ 매일 오름</p><noscript data-reactid=".0.0.0.3.4.$763650660.0.1:$icon-763650660"></noscript><div class="box_reply" data-reactid=".0.0.0.3.4.$763650660.0.3"><span class="box_inner" data-reactid=".0.0.0.3.4.$763650660.0.3.0"><button class="reply_count #reply ?c_title=%EB%8B%B5%EA%B8%80" aria-expended="false" data-tiara-action-name="답글목록_열기" data-reactid=".0.0.0.3.4.$763650660.0.3.0.0"><span data-reactid=".0.0.0.3.4.$763650660.0.3.0.0.0">답글 작성</span></button></span><span class="comment_recomm" data-reactid=".0.0.0.3.4.$763650660.0.3.1"><button class="btn_g btn_recomm #like ?c_title=%EB%8C%93%EA%B8%80%EC%B0%AC%EC%84%B1" data-tiara-action-name="댓글_찬성" data-tiara-custom="alex.commentId=763650660" data-reactid=".0.0.0.3.4.$763650660.0.3.1.0"><span class="img_cmt ico_recomm" data-reactid=".0.0.0.3.4.$763650660.0.3.1.0.0">댓글 찬성하기</span><span class="num_txt" data-reactid=".0.0.0.3.4.$763650660.0.3.1.0.1">0</span></button><button class="btn_g btn_oppose #dislike ?c_title=%EB%8C%93%EA%B8%80%EB%B0%98%EB%8C%80" data-tiara-action-name="댓글_반대" data-tiara-custom="alex.commentId=763650660" data-reactid=".0.0.0.3.4.$763650660.0.3.1.1"><span class="img_cmt ico_oppose" data-reactid=".0.0.0.3.4.$763650660.0.3.1.1.0">댓글 비추천하기</span><span class="num_txt" data-reactid=".0.0.0.3.4.$763650660.0.3.1.1.1">0</span></button></span></div><a href="javascript:" class="hide_btn" data-tiara-action-name="댓글_덮기" data-tiara-custom="alex.commentId=763650660" data-reactid=".0.0.0.3.4.$763650660.0.5"><span class="img_cmt ico_hide" data-reactid=".0.0.0.3.4.$763650660.0.5.0"><span class="screen_out" data-reactid=".0.0.0.3.4.$763650660.0.5.0.0">댓글 덮기</span></span></a><button href="javascript:" class="tooltip" aria-expanded="false" data-reactid=".0.0.0.3.4.$763650660.0.6"><span class="img_cmt tooltip_ico" data-reactid=".0.0.0.3.4.$763650660.0.6.0"><span class="screen_out" data-reactid=".0.0.0.3.4.$763650660.0.6.0.0">댓글 옵션 버튼 펼치기</span></span></button></div></li><li id="comment763625107" class="" data-reactid=".0.0.0.3.4.$763625107"><div class="cmt_info" data-reactid=".0.0.0.3.4.$763625107.0"><strong class="tit_nick" data-reactid=".0.0.0.3.4.$763625107.0.0"><noscript data-reactid=".0.0.0.3.4.$763625107.0.0.0"></noscript><span class="info_author" data-reactid=".0.0.0.3.4.$763625107.0.0.1"><span class="link_nick" data-reactid=".0.0.0.3.4.$763625107.0.0.1.0"><span class="img_cmt img_daum" data-reactid=".0.0.0.3.4.$763625107.0.0.1.0.0"></span></span><a role="button" href="javascript:" class="link_nick" data-userid="60872075" data-tiara-action-name="다른사람댓글활동_보기" data-reactid=".0.0.0.3.4.$763625107.0.0.1.1"><span class="screen_out" data-reactid=".0.0.0.3.4.$763625107.0.0.1.1.0">댓글 작성자</span><span data-reactid=".0.0.0.3.4.$763625107.0.0.1.1.1">fdaaf</span><span class="screen_out" data-reactid=".0.0.0.3.4.$763625107.0.0.1.1.2">, 댓글모아보기</span></a><span class="txt_date" data-reactid=".0.0.0.3.4.$763625107.0.0.1.2">3시간전</span></span></strong><p class="desc_txt font_size_17" data-reactid=".0.0.0.3.4.$763625107.0.1:$desc-763625107">국제유가 오르면 정유사는 떼돈번다</p><noscript data-reactid=".0.0.0.3.4.$763625107.0.1:$icon-763625107"></noscript><div class="box_reply" data-reactid=".0.0.0.3.4.$763625107.0.3"><span class="box_inner" data-reactid=".0.0.0.3.4.$763625107.0.3.0"><button class="reply_count #reply ?c_title=%EB%8B%B5%EA%B8%80" aria-expended="false" data-tiara-action-name="답글목록_열기" data-reactid=".0.0.0.3.4.$763625107.0.3.0.0"><span data-reactid=".0.0.0.3.4.$763625107.0.3.0.0.0">답글 작성</span></button></span><span class="comment_recomm" data-reactid=".0.0.0.3.4.$763625107.0.3.1"><button class="btn_g btn_recomm #like ?c_title=%EB%8C%93%EA%B8%80%EC%B0%AC%EC%84%B1" data-tiara-action-name="댓글_찬성" data-tiara-custom="alex.commentId=763625107" data-reactid=".0.0.0.3.4.$763625107.0.3.1.0"><span class="img_cmt ico_recomm" data-reactid=".0.0.0.3.4.$763625107.0.3.1.0.0">댓글 찬성하기</span><span class="num_txt" data-reactid=".0.0.0.3.4.$763625107.0.3.1.0.1">1</span></button><button class="btn_g btn_oppose #dislike ?c_title=%EB%8C%93%EA%B8%80%EB%B0%98%EB%8C%80" data-tiara-action-name="댓글_반대" data-tiara-custom="alex.commentId=763625107" data-reactid=".0.0.0.3.4.$763625107.0.3.1.1"><span class="img_cmt ico_oppose" data-reactid=".0.0.0.3.4.$763625107.0.3.1.1.0">댓글 비추천하기</span><span class="num_txt" data-reactid=".0.0.0.3.4.$763625107.0.3.1.1.1">0</span></button></span></div><a href="javascript:" class="hide_btn" data-tiara-action-name="댓글_덮기" data-tiara-custom="alex.commentId=763625107" data-reactid=".0.0.0.3.4.$763625107.0.5"><span class="img_cmt ico_hide" data-reactid=".0.0.0.3.4.$763625107.0.5.0"><span class="screen_out" data-reactid=".0.0.0.3.4.$763625107.0.5.0.0">댓글 덮기</span></span></a><button href="javascript:" class="tooltip" aria-expanded="false" data-reactid=".0.0.0.3.4.$763625107.0.6"><span class="img_cmt tooltip_ico" data-reactid=".0.0.0.3.4.$763625107.0.6.0"><span class="screen_out" data-reactid=".0.0.0.3.4.$763625107.0.6.0.0">댓글 옵션 버튼 펼치기</span></span></button></div></li></ul><div class="more_report" data-reactid=".0.0.0.3.6"><a class="link_fold fst" data-reactid=".0.0.0.3.6.0"><span class="img_cmt ico_alex" data-reactid=".0.0.0.3.6.0.0"></span><span data-reactid=".0.0.0.3.6.0.1">본문보기</span></a><a class="link_fold write_reply_btn" data-reactid=".0.0.0.3.6.1"><span class="img_cmt ico_cmt" data-reactid=".0.0.0.3.6.1.0"></span><span data-reactid=".0.0.0.3.6.1.1">댓글 쓰기</span></a></div></div></div></div></div></div>    
       </div> 
      </div> 
     </div> 
     <div id="mAside" data-cloud-area="aside" style="position: static; top: inherit; left: inherit;"> 
      <ins style="display: block; width: 250px; height: 250px; text-decoration: none; max-height: inherit; min-height: 250px; min-width: 250px;" class="kakao_ad_area aside_g aside_ad" data-ad-unit="05d24" data-ad-type="D" data-ad-width="250" data-ad-height="250" data-cloud="pc_aside_top_ad" data-ad-param-channel="harmony" data-ad-param-cp="11_pc_media_news" id="kakao_ad_qvizBi"><div data-ad-creative-wrap="outer" class="adfit__html-banner" style="min-width: 250px; min-height: 250px;"><div data-ad-creative-wrap="inner" class="adfit__html-banner__inner" style="height: 250px;"><div><iframe src="https://t1.daumcdn.net/adfit/adunit_style/88b0a7786a81c005b0bb72885df94004ed059b45" width="100%" style="position: absolute; top: 0px; left: 0px; display: block; border: none; height: 250px; width: 250px; min-width: 250px; min-height: 250px; overflow: hidden; padding: 0px; margin: 0px auto;" frameborder="0" scrolling="no" marginheight="0" marginwidth="0" name="%5B%7B%22usePrefersColorScheme%22%3A%22N%22%2C%22usePreferColorScheme%22%3A%22N%22%2C%22frameId%22%3A%2232bafd6c-58cb-4e37-9d9f-c7794ccd9c5f%22%2C%22frameName%22%3A%2232bafd6c-58cb-4e37-9d9f-c7794ccd9c5f%22%2C%22title%22%3A%22%EC%89%BF!%20%ED%95%98%ED%95%84%20%EC%A7%80%EA%B8%88%20%ED%95%A0%EC%9D%B8%20%EC%A4%91%22%2C%22mainImage%22%3A%7B%22url%22%3A%22https%3A%2F%2Ft1.daumcdn.net%2Fb2%2Fcreative%2F338592%2F928f27ae85e06904d1fc8146ba151f76.jpg%22%2C%22width%22%3A1200%2C%22height%22%3A600%7D%2C%22profileName%22%3A%22%EC%8D%A8%EB%B8%8C%EC%9B%A8%EC%9D%B4%22%2C%22adInfoIcon%22%3A%7B%22url%22%3A%22https%3A%2F%2Ft1.daumcdn.net%2Fbiz%2Fui%2Fad%2FADmark%2Fi_mark_200803.png%22%2C%22width%22%3A15%2C%22height%22%3A15%7D%2C%22adInfoUrl%22%3A%22https%3A%2F%2Finfo.ad.daum.net%2Foptout.do%22%2C%22landingUrl%22%3A%22https%3A%2F%2Ftr.ad.daum.net%2Fclk%3Fwa%3D8Utrs9widOeytlj2n2eKTA%26enc%3DhRyvjkrkjnmxE_Rr4-l-pda1a1mzuVVL84G1yin39RDbrcE3aC-NcFpsPM6LvrvJfUvYhkrAH5bpCVZFbTt2st3ow3-L-SgvRPjgS2Hve7pE8A7HS8AEoN_w4XvIPXGkVvfeZ-WtMxndDIt3qK4EX2kOvKw759DHsraMQ6xGoUjN9_wy9zOvVQD06DBnG36cZGFH0bs2ImuZIoR3Me8UWaAEBKYIet2fwzKrL9o6aC1g_7o5IAtsuAvAMPnmspvAGnAVetJOv1oP_uANolW9TvP4vMcHFMknWuvXwRChXQ0FMyJmbK4dIsQZAbM7_H5MxbGT94OmQdoqmN0fF2OUMZ4m7M_ka8AgTVOSRSkIsPrdWBmOYghuQ0zvLVe23t8VoSkx0wfwX7rVPOXPlfsp39Kj7zyC_2aCYmExAKd2QjQ%22%2C%22useAdInfoIcon%22%3Afalse%2C%22useAdInfoUrl%22%3Afalse%2C%22events%22%3A%5B%7B%22type%22%3A%22viewable%22%2C%22url%22%3A%22https%3A%2F%2Ftr.ad.daum.net%2Fvimp%3Fwa%3D8Utrs9widOeytlj2n2eKTA%26enc%3DhRyvjkrkjnmxE_Rr4-l-pda1a1mzuVVL84G1yin39RDbrcE3aC-NcFpsPM6LvrvJfUvYhkrAH5bpCVZFbTt2st3ow3-L-SgvRPjgS2Hve7pE8A7HS8AEoN_w4XvIPXGkVvfeZ-WtMxndDIt3qK4EX2kOvKw759DHsraMQ6xGoUjN9_wy9zOvVQD06DBnG36cZGFH0bs2ImuZIoR3Me8UWaAEBKYIet2fwzKrL9o6aC1g_7o5IAtsuAvAMPnmspvAN6jsUGnbxq3sgfpqUDl2nXx9l8TdRCuvmVmjnR5Z8bMQAwVsO06K10X9ao81VImO6qShqgomCH5ECF8_H901Yw%22%7D%2C%7B%22type%22%3A%22hide%22%2C%22url%22%3A%22https%3A%2F%2Ftr.ad.daum.net%2Fac%3Fwa%3D8Utrs9widOeytlj2n2eKTA%26enc%3DhRyvjkrkjnmxE_Rr4-l-pda1a1mzuVVL84G1yin39RDbrcE3aC-NcFpsPM6LvrvJfUvYhkrAH5bpCVZFbTt2st3ow3-L-SgvRPjgS2Hve7pE8A7HS8AEoN_w4XvIPXGkVvfeZ-WtMxndDIt3qK4EX2kOvKw759DHsraMQ6xGoUjN9_wy9zOvVQD06DBnG36cZGFH0bs2ImuZIoR3Me8UWaAEBKYIet2fwzKrL9o6aC1g_7o5IAtsuAvAMPnmspvAN6jsUGnbxq3sgfpqUDl2nXx9l8TdRCuvmVmjnR5Z8bMQAwVsO06K10X9ao81VImO2J-16rGwJqIqSC0xm-TrIg%22%7D%2C%7B%22type%22%3A%22rendered%22%2C%22url%22%3A%22https%3A%2F%2Fkyson.ad.daum.net%2Fad%2Frendered%3Frequest_id%3D430030b8-a09d-4676-9f9a-e31be0e5fd09%26adunit_id%3D05d24%26bid_id%3D20220228021941_G8AVb%26dsp_id%3DMOMENT%26med_dsp_id%3DMOMENT%26network_type%3DDAUM%26sdk_type%3Dweb%26sdk_version%3D4.1.5%26ad_type%3DBanner%26w%3D250%26h%3D250%26template_seq%3D15%22%7D%2C%7B%22type%22%3A%22viewable%22%2C%22url%22%3A%22https%3A%2F%2Fkyson.ad.daum.net%2Fad%2Fviewable%3Frequest_id%3D430030b8-a09d-4676-9f9a-e31be0e5fd09%26adunit_id%3D05d24%26bid_id%3D20220228021941_G8AVb%26dsp_id%3DMOMENT%26med_dsp_id%3DMOMENT%26network_type%3DDAUM%26sdk_type%3Dweb%26sdk_version%3D4.1.5%26ad_type%3DBanner%26w%3D250%26h%3D250%26template_seq%3D15%22%7D%2C%7B%22type%22%3A%22click%22%2C%22url%22%3A%22https%3A%2F%2Fkyson.ad.daum.net%2Fad%2Fclick%3Frequest_id%3D430030b8-a09d-4676-9f9a-e31be0e5fd09%26adunit_id%3D05d24%26bid_id%3D20220228021941_G8AVb%26dsp_id%3DMOMENT%26med_dsp_id%3DMOMENT%26network_type%3DDAUM%26sdk_type%3Dweb%26sdk_version%3D4.1.5%26ad_type%3DBanner%26w%3D250%26h%3D250%26template_seq%3D15%22%7D%2C%7B%22type%22%3A%22duration%22%2C%22url%22%3A%22https%3A%2F%2Fkyson.ad.daum.net%2Fad%2Fduration%3Frequest_id%3D430030b8-a09d-4676-9f9a-e31be0e5fd09%26adunit_id%3D05d24%26bid_id%3D20220228021941_G8AVb%26dsp_id%3DMOMENT%26med_dsp_id%3DMOMENT%26network_type%3DDAUM%26sdk_type%3Dweb%26sdk_version%3D4.1.5%26ad_type%3DBanner%26w%3D250%26h%3D250%26template_seq%3D15%22%7D%5D%2C%22dspId%22%3A%22MOMENT%22%2C%22altText%22%3A%22%EC%8D%A8%EB%B8%8C%EC%9B%A8%EC%9D%B4%20%EC%9D%B4%ED%83%88%EB%A6%AC%EC%95%88%20%EB%B9%84%EC%97%A0%ED%8B%B0%22%2C%22nasTemplateSeq%22%3A15%2C%22type%22%3A%22native%22%7D%5D" id="32bafd6c-58cb-4e37-9d9f-c7794ccd9c5f"></iframe></div></div><div class="adfit__border adfit__border--left" data-ad-creative-wrap="border" style="background-color: rgb(229, 229, 229);"></div><div class="adfit__border adfit__border--right" data-ad-creative-wrap="border" style="background-color: rgb(229, 229, 229);"></div><div class="adfit__border adfit__border--top" data-ad-creative-wrap="border" style="background-color: rgb(229, 229, 229);"></div><div class="adfit__border adfit__border--bottom" data-ad-creative-wrap="border" style="background-color: rgb(229, 229, 229);"></div></div> </ins> 
      <div class="aside_g aside_phototv" data-cloud="pc_news_photo_tv"> 
       <h3 class="txt_newsview">포토&amp;TV</h3> 
       <div class="cont_aside"> 
        <ul class="list_phototv"><li>
 <a class="link_cont" href="https://gallery.v.daum.net/p/viewer/1518/20220228002701935">
 <span class="wrap_thumb">
 <img class="thumb_g" src="https://img1.daumcdn.net/thumb/S144x140/?fname=https://t1.daumcdn.net/news/202202/28/sbsi/20220228002705211udva.jpg" alt="">
 <span class="ico_phototv">
 
 <span class="ico_newsview ico_photo">포토</span>
 
 </span>
 </span>
 <strong class="tit_thumb">대구 달성군 주암산서 다시 발화..산불진화대원 90명 투입</strong>
 </a>
 </li><li>
 <a class="link_cont" href="https://news.v.daum.net/v/20220227233707433?s=tv_news">
 <span class="wrap_thumb">
 <img class="thumb_g" src="https://img1.daumcdn.net/thumb/S144x140/?fname=https://t1.daumcdn.net/news/202202/27/YTN/20220227233708639dxyq.jpg" alt="">
 <span class="ico_phototv">
 
 <span class="ico_newsview ico_vod">영상</span>
 
 </span>
 </span>
 <strong class="tit_thumb">[날씨] 내일 구름 많고 낮 포근..큰 일교차 주의</strong>
 </a>
 </li></ul> 
       </div> 
      </div>  
      <ins style="display: block; width: 250px; height: 250px; text-decoration: none; max-height: inherit; min-height: 250px; min-width: 250px; position: relative;" class="kakao_ad_area aside_g aside_ad" data-ad-unit="0QH44" data-ad-type="D" data-ad-width="250" data-ad-height="250" data-cloud="pc_aside_bottom_ad" data-ad-param-channel="harmony" data-ad-param-cp="11_pc_media_news" id="kakao_ad_QAHaw5"><div id="sf_pos_rel_el_sf_pos_kakao_ad_QAHaw5" class="iab_sf" style="position: relative; z-index: 0; width: 250px; height: 250px; visibility: inherit; min-width: 250px; min-height: 250px;"><iframe style="position: absolute; z-index: 0; width: 250px; height: 250px; top: 0px; left: 0px; visibility: inherit; display: block; max-width: none; max-height: none; min-width: 250px; min-height: 250px;" id="sf_frm_kakao_ad_QAHaw5" src="https://t1.daumcdn.net/kas/static/third-party/sf/1-1-1/html/r.html" frameborder="no" scrolling="no" allowtransparency="true" hidefocus="true" tabindex="-1" marginwidth="0" marginheight="0" title="광고"></iframe></div><div class="adfit__border adfit__border--left" data-ad-creative-wrap="border" style="background-color: rgb(229, 229, 229);"></div><div class="adfit__border adfit__border--right" data-ad-creative-wrap="border" style="background-color: rgb(229, 229, 229);"></div><div class="adfit__border adfit__border--top" data-ad-creative-wrap="border" style="background-color: rgb(229, 229, 229);"></div><div class="adfit__border adfit__border--bottom" data-ad-creative-wrap="border" style="background-color: rgb(229, 229, 229);"></div> </ins> 
     </div> 
    </div> 
    <div id="cEtc"> 
     <div class="realtime_view"> 
      <div class="inner_realtime"> 
       <div data-cloud-area="etc"> 
        <div class="realtime_news" data-cloud="pc_realtime_news"> 
         <h3 class="txt_newsview">이 시각 추천뉴스</h3> 
         <ul class="list_realtime"> 
          <li class="item_main"><a href="https://v.daum.net/v/20220227233602413" class="link_cont" data-itemid="20220227233602413" "="" data-pos="">
 <span class="wrap_thumb"><img src="//img2.daumcdn.net/thumb/S570x320.fwebp/?fname=https://t1.daumcdn.net/news/202202/27/seoul/20220227233603533zaxv.jpg" class="thumb_g" alt=""></span>
 <strong class="tit_thumb">"우크라 침공 계획보다 더뎌.. 푸틴, 노발대발"</strong>
 </a></li> 
          <li class="item_main"><a href="https://v.daum.net/v/20220227234800505" class="link_cont" data-itemid="20220227234800505" "="" data-pos="">
 <span class="wrap_thumb"><img src="//img1.daumcdn.net/thumb/S570x320.fwebp/?fname=https://t1.daumcdn.net/news/202202/27/ohmynews/20220227234803110txyo.jpg" class="thumb_g" alt=""></span>
 <strong class="tit_thumb">틀린 사실, 막말까지.. 한층 더 거칠어진 윤석열</strong>
 </a></li> 
          <li><strong class="tit_g">
 <span class="info_news">조선일보</span>
 <a href="https://v.daum.net/v/20220227041956693" data-itemid="20220227041956693" data-pos="2" class="link_txt">"과거 침략당해 원조 받아놓고.." 韓 러시아 제재 지연에 쏟아진 비판</a>
 </strong></li> 
          <li><strong class="tit_g">
 <span class="info_news">중앙일보</span>
 <a href="https://v.daum.net/v/20220227144843920" data-itemid="20220227144843920" data-pos="3" class="link_txt">바이든 "러시아 제재 함께한 동료들" 거론..한국은 또 빠졌다</a>
 </strong></li> 
          <li><strong class="tit_g">
 <span class="info_news">뉴시스</span>
 <a href="https://v.daum.net/v/20220227184936243" data-itemid="20220227184936243" data-pos="4" class="link_txt">與 "윤석열 유세장서 또 여성 폭행..야만적 행태 개탄"</a>
 </strong></li> 
          <li><strong class="tit_g">
 <span class="info_news">뉴스1</span>
 <a href="https://v.daum.net/v/20220227190000397" data-itemid="20220227190000397" data-pos="5" class="link_txt">[영상] 연료 바닥나고 길 잃은 러시아 군인들에게 우크라이나 운전자가 던진 말</a>
 </strong></li> 
          <li><strong class="tit_g">
 <span class="info_news">서울신문</span>
 <a href="https://v.daum.net/v/20220227140603154" data-itemid="20220227140603154" data-pos="6" class="link_txt">日언론 "푸틴을 이렇게 만든 것은 아베"..우크라 침공 책임론 제기</a>
 </strong></li> 
          <li><strong class="tit_g">
 <span class="info_news">세계일보</span>
 <a href="https://v.daum.net/v/20220227220129477" data-itemid="20220227220129477" data-pos="7" class="link_txt">"K-방산 잇단 '대박'.. 中企 함께 수출길 뚫어야" [차 한잔 나누며]</a>
 </strong></li> 
          <li><strong class="tit_g">
 <span class="info_news">세계일보</span>
 <a href="https://v.daum.net/v/20220227220132480" data-itemid="20220227220132480" data-pos="8" class="link_txt">사상 최대 실적에도.. 워런 버핏 "투자할 곳 없어"</a>
 </strong></li> 
          <li><strong class="tit_g">
 <span class="info_news">쿠키뉴스</span>
 <a href="https://v.daum.net/v/20220227201802176" data-itemid="20220227201802176" data-pos="9" class="link_txt">"대게 10마리 먹은 듯".. 윤석열, 단일화 결렬 후 포항서 되찾은 '미소'</a>
 </strong></li> 
          <li><strong class="tit_g">
 <span class="info_news">YTN</span>
 <a href="https://v.daum.net/v/20220227223001854" data-itemid="20220227223001854" data-pos="10" class="link_txt">"한국 도움 절실"..재한 우크라·러시아인 '전쟁반대' 집회</a>
 </strong></li> 
          <li><strong class="tit_g">
 <span class="info_news">중앙일보</span>
 <a href="https://v.daum.net/v/20220228000250669" data-itemid="20220228000250669" data-pos="11" class="link_txt">우크라이나 사태가 불붙인 기름값, 물가 4%대 상승 우려</a>
 </strong></li> 
         </ul> 
        </div>  
       </div> 
       <div class="realtime_aside" id="realtime_aside" data-cloud-area="etc-aside"> 
        <ins class="kakao_ad_area etc_bottom_ad" style="display: block; width: 250px; margin: 0px auto; text-decoration: none; height: 250px; max-height: inherit; min-height: 250px; min-width: 250px;" data-ad-unit="DAN-kQxraZPBvpl52CWq" data-ad-width="250" data-ad-height="250" data-ad-onfail="onFailEtcBottomAd" data-cloud="pc_etc_bottom_ad" data-ad-param-channel="harmony" data-ad-param-cp="11_pc_media_news" id="kakao_ad_a90LSz"><div data-ad-creative-wrap="outer" class="adfit__html-banner" style="min-width: 250px; min-height: 250px;"><div data-ad-creative-wrap="inner" class="adfit__html-banner__inner" style="height: 250px;"><div><iframe src="https://t1.daumcdn.net/adfit/adunit_style/6ab5ec5b821bc9bc6aad6210feec8add27278b51" width="100%" style="position: absolute; top: 0px; left: 0px; display: block; border: none; height: 250px; width: 250px; min-width: 250px; min-height: 250px; overflow: hidden; padding: 0px; margin: 0px auto;" frameborder="0" scrolling="no" marginheight="0" marginwidth="0" name="%5B%7B%22usePrefersColorScheme%22%3A%22N%22%2C%22usePreferColorScheme%22%3A%22N%22%2C%22frameId%22%3A%22177505bc-5f67-469c-a488-15261f2d34e6%22%2C%22frameName%22%3A%22177505bc-5f67-469c-a488-15261f2d34e6%22%2C%22title%22%3A%223%EC%9B%945~6%EC%9D%BC%20%EB%8C%80%EA%B5%AC%EC%9B%A8%EB%94%A9%EB%B0%95%EB%9E%8C%ED%9A%8C%22%2C%22body%22%3A%22in%20EXCO%EC%9D%B8%ED%84%B0%EB%B6%88%EA%B3%A0%ED%98%B8%ED%85%94%2F%EC%9B%A8%EB%94%A9%ED%99%80%2C%EC%8A%A4%EB%93%9C%EB%A9%94%20%ED%8A%B9%EA%B0%80%ED%95%A0%EC%9D%B8%26%EA%B3%84%EC%95%BD%EA%B8%88%EC%A7%80%EC%9B%90%2F%20%EC%84%A0%EC%B0%A9%EC%88%9C%2C%EB%B0%A9%EB%AC%B8%EC%9D%B4%EB%B2%A4%ED%8A%B8%22%2C%22mainImage%22%3A%7B%22url%22%3A%22https%3A%2F%2Fimg1.daumcdn.net%2Fthumb%2FS500x500%2F%3Ffname%3Dhttps%3A%2F%2Ft1.kakaocdn.net%2Fkakao_ka_public%2Fcreative%2F362517-b5363491-66a1-4468-bc47-abc1b26edade.jpg%26scode%3Db2_01%22%2C%22width%22%3A500%2C%22height%22%3A500%2C%22link%22%3A%7B%22url%22%3A%22https%3A%2F%2Fka.ad.daum.net%2Fclick%2Fhttp%253A%252F%252F%25EB%258C%2580%25EA%25B5%25AC%25EC%259B%25A8%25EB%2594%25A9%25EB%25B0%2595%25EB%259E%258C%25ED%259A%258C.kr%253Futm_source%253Dkakao%2526utm_medium%253Dkeyword%2526utm_campaign%253DPREMIUM_LINK%2526k_campaign%253D217249763460550656%2526k_adgroup%253D217249899193241600%2526k_media%253DPREMIUM_LINK_CONTENT%2526k_creative%253D282586362981294081%2526k_creativelink%253D282586363048402945%2526k_rank%253D1%2FTpn9jrMaHHPsWmxn2b6D508f20huXKe_wyvwWNbLMww12NYrIAww7wq3Dd8PF-rSg4HE0CbVdeKf2dZdOgs3Prt4wTzs_OQEpUC8vKLdm0aobc7k45RKfQo20bFKDjq9m9Y2WW2HFZixlJKcM2YEy42BUDzP7_a8P9o4bX3AfTZD38q3TeMZlPtrGx6GclpHEKH563A1vlpfA0LZJl1OeLt_IRgWzMQdPChatr-By2xakfSHMVW_yGSfBVxmpguLFyYoc1piMBoA5p48OXjIoA_xKQrOooNl434PaliNbZeXHSJ2j6mHw6ASV_1SFdkegcqtQc0w8AOf9P5JqHjQ7vGvm5Se59ZJp8wLx7Zm1wtIXNRtmoKW0YN3bK9EKWOdfOchGznkkqmiNFUhOJyRGntqCpo78MqZLmNHz70xqqYffb-tQEuQInkONS4FOZKHPYG-CtKTM9jye1n81dE%3D%3Fp%3DLuibGa8R3NggtP1YaqB869-uHOkhNjIipK7AVyBTdTQ%22%7D%7D%2C%22adInfoIcon%22%3A%7B%22url%22%3A%22https%3A%2F%2Ft1.daumcdn.net%2Fbiz%2Fui%2Fad%2FADmark%2Fi_mark_200803.png%22%2C%22width%22%3A15%2C%22height%22%3A15%7D%2C%22adInfoUrl%22%3A%22https%3A%2F%2Finfo.ad.daum.net%2Foptout.do%22%2C%22landingUrl%22%3A%22https%3A%2F%2Fka.ad.daum.net%2Fclick%2Fhttp%253A%252F%252F%25EB%258C%2580%25EA%25B5%25AC%25EC%259B%25A8%25EB%2594%25A9%25EB%25B0%2595%25EB%259E%258C%25ED%259A%258C.kr%253Futm_source%253Dkakao%2526utm_medium%253Dkeyword%2526utm_campaign%253DPREMIUM_LINK%2526k_campaign%253D217249763460550656%2526k_adgroup%253D217249899193241600%2526k_media%253DPREMIUM_LINK_CONTENT%2526k_creative%253D282586362981294081%2526k_creativelink%253D282586363048402945%2526k_rank%253D1%2FTpn9jrMaHHPsWmxn2b6D508f20huXKe_wyvwWNbLMww12NYrIAww7wq3Dd8PF-rSg4HE0CbVdeKf2dZdOgs3Prt4wTzs_OQEpUC8vKLdm0aobc7k45RKfQo20bFKDjq9m9Y2WW2HFZixlJKcM2YEy42BUDzP7_a8P9o4bX3AfTZD38q3TeMZlPtrGx6GclpHEKH563A1vlpfA0LZJl1OeLt_IRgWzMQdPChatr-By2xakfSHMVW_yGSfBVxmpguLFyYoc1piMBoA5p48OXjIoA_xKQrOooNl434PaliNbZeXHSJ2j6mHw6ASV_1SFdkegcqtQc0w8AOf9P5JqHjQ7vGvm5Se59ZJp8wLx7Zm1wtIXNRtmoKW0YN3bK8VVn0%3D%3Fp%3DLuibGa8R3NggtP1YaqB869-uHOkhNjIipK7AVyBTdTQ%22%2C%22useAdInfoIcon%22%3Afalse%2C%22useAdInfoUrl%22%3Afalse%2C%22events%22%3A%5B%7B%22type%22%3A%22viewable%22%2C%22url%22%3A%22https%3A%2F%2Fka.ad.daum.net%2Fvimp%2FTpn9jrMaHHPsWmxn2b6D508f20huXKe_wyvwWNbLMww12NYrIAww7wq3Dd8PF-rSg4HE0CbVdeKf2dZdOgs3Prt4wTzs_OQEpUC8vKLdm0aobc7k45RKfQo20bFKDjq9m9Y2WW2HFZixlJKcM2YEy42BUDzP_Pe-aIUQYn2IZ3pcmIPyEoQC%3Fp%3DLuibGa8R3NggtP1YaqB869-uHOkhNjIipK7AVyBTdTQ%22%7D%2C%7B%22type%22%3A%22rendered%22%2C%22url%22%3A%22https%3A%2F%2Fka.ad.daum.net%2Frimp%2FTpn9jrMaHHPsWmxn2b6D508f20huXKe_wyvwWNbLMww12NYrIAww7wq3Dd8PF-rSg4HE0CbVdeKf2dZdOgs3Prt4wTzs_OQEpUC8vKLdm0aobc7k45RKfQo20bFKDjq9m9Y2WW2HFZixlJKcM2YEy42BUDzP_Pe-aIUQYn2IZ3pcmIPyEoQC%3Fp%3DLuibGa8R3NggtP1YaqB869-uHOkhNjIipK7AVyBTdTQ%26ac%3DA4-sguV9QDCuDyxxpO_C5VUWixg5XqW9kmD3CNfHKlkqiNIyLFBjuxHnA4gLGaLQ%22%7D%2C%7B%22type%22%3A%22rendered%22%2C%22url%22%3A%22https%3A%2F%2Fkyson.ad.daum.net%2Fad%2Frendered%3Frequest_id%3D8d3136ca-e1b9-43a2-9edf-a9d17d2e46da%26adunit_id%3DDAN-kQxraZPBvpl52CWq%26bid_id%3Dbd995535-6d59-448b-bfc4-5a9d17ff60b01%26dsp_id%3DKEYWORDAD%26med_dsp_id%3DKAAS_CM%26network_type%3DDAUM%26sdk_type%3Dweb%26sdk_version%3D4.1.5%26ad_type%3DBanner%26w%3D250%26h%3D250%26template_seq%3D34%22%7D%2C%7B%22type%22%3A%22viewable%22%2C%22url%22%3A%22https%3A%2F%2Fkyson.ad.daum.net%2Fad%2Fviewable%3Frequest_id%3D8d3136ca-e1b9-43a2-9edf-a9d17d2e46da%26adunit_id%3DDAN-kQxraZPBvpl52CWq%26bid_id%3Dbd995535-6d59-448b-bfc4-5a9d17ff60b01%26dsp_id%3DKEYWORDAD%26med_dsp_id%3DKAAS_CM%26network_type%3DDAUM%26sdk_type%3Dweb%26sdk_version%3D4.1.5%26ad_type%3DBanner%26w%3D250%26h%3D250%26template_seq%3D34%22%7D%2C%7B%22type%22%3A%22click%22%2C%22url%22%3A%22https%3A%2F%2Fkyson.ad.daum.net%2Fad%2Fclick%3Frequest_id%3D8d3136ca-e1b9-43a2-9edf-a9d17d2e46da%26adunit_id%3DDAN-kQxraZPBvpl52CWq%26bid_id%3Dbd995535-6d59-448b-bfc4-5a9d17ff60b01%26dsp_id%3DKEYWORDAD%26med_dsp_id%3DKAAS_CM%26network_type%3DDAUM%26sdk_type%3Dweb%26sdk_version%3D4.1.5%26ad_type%3DBanner%26w%3D250%26h%3D250%26template_seq%3D34%22%7D%2C%7B%22type%22%3A%22duration%22%2C%22url%22%3A%22https%3A%2F%2Fkyson.ad.daum.net%2Fad%2Fduration%3Frequest_id%3D8d3136ca-e1b9-43a2-9edf-a9d17d2e46da%26adunit_id%3DDAN-kQxraZPBvpl52CWq%26bid_id%3Dbd995535-6d59-448b-bfc4-5a9d17ff60b01%26dsp_id%3DKEYWORDAD%26med_dsp_id%3DKAAS_CM%26network_type%3DDAUM%26sdk_type%3Dweb%26sdk_version%3D4.1.5%26ad_type%3DBanner%26w%3D250%26h%3D250%26template_seq%3D34%22%7D%5D%2C%22dspId%22%3A%22KEYWORDAD%22%2C%22bodyLink%22%3A%7B%22url%22%3A%22https%3A%2F%2Fka.ad.daum.net%2Fclick%2Fhttp%253A%252F%252F%25EB%258C%2580%25EA%25B5%25AC%25EC%259B%25A8%25EB%2594%25A9%25EB%25B0%2595%25EB%259E%258C%25ED%259A%258C.kr%253Futm_source%253Dkakao%2526utm_medium%253Dkeyword%2526utm_campaign%253DPREMIUM_LINK%2526k_campaign%253D217249763460550656%2526k_adgroup%253D217249899193241600%2526k_media%253DPREMIUM_LINK_CONTENT%2526k_creative%253D282586362981294081%2526k_creativelink%253D282586363048402945%2526k_rank%253D1%2FTpn9jrMaHHPsWmxn2b6D508f20huXKe_wyvwWNbLMww12NYrIAww7wq3Dd8PF-rSg4HE0CbVdeKf2dZdOgs3Prt4wTzs_OQEpUC8vKLdm0aobc7k45RKfQo20bFKDjq9m9Y2WW2HFZixlJKcM2YEy42BUDzP7_a8P9o4bX3AfTZD38q3TeMZlPtrGx6GclpHEKH563A1vlpfA0LZJl1OeLt_IRgWzMQdPChatr-By2xakfSHMVW_yGSfBVxmpguLFyYoc1piMBoA5p48OXjIoA_xKQrOooNl434PaliNbZeXHSJ2j6mHw6ASV_1SFdkegcqtQc0w8AOf9P5JqHjQ7vGvm5Se59ZJp8wLx7Zm1wtIXNRtmoKW0YN3bK9EKWOdfOchGznkkqmiNEUsPpKBHWp3D_dZ8Na3IA%3D%3D%3Fp%3DLuibGa8R3NggtP1YaqB869-uHOkhNjIipK7AVyBTdTQ%22%7D%2C%22nasTemplateSeq%22%3A34%2C%22type%22%3A%22native%22%7D%5D" id="177505bc-5f67-469c-a488-15261f2d34e6"></iframe></div></div><div class="adfit__border adfit__border--left" data-ad-creative-wrap="border" style="background-color: rgb(229, 229, 229);"></div><div class="adfit__border adfit__border--right" data-ad-creative-wrap="border" style="background-color: rgb(229, 229, 229);"></div><div class="adfit__border adfit__border--top" data-ad-creative-wrap="border" style="background-color: rgb(229, 229, 229);"></div><div class="adfit__border adfit__border--bottom" data-ad-creative-wrap="border" style="background-color: rgb(229, 229, 229);"></div></div> </ins>  
       </div> 
      </div> 
     </div> 
     <div data-cloud-area="etc-addition"> 
     </div> 
    </div> 
   </div> 
   <hr class="hide"> 
   <div class="footer_comm" id="kakaoFoot" role="contentinfo" data-cloud-area="footer"> 
    <div class="inner_foot" data-cloud="pc_footer"> 
     <div class="direct_link"> 
      <h2 class="tit_direct">바로가기</h2> 
      <div class="item_bundle"> <a class="link_txt" href="https://news.daum.net/">뉴스홈</a> <a class="link_txt" href="https://news.daum.net/society/">사회</a> <a class="link_txt" href="https://news.daum.net/politics/">정치</a> <a class="link_txt" href="https://news.daum.net/economic/">경제</a> <a class="link_txt" href="https://news.daum.net/foreign/">국제</a> <a class="link_txt" href="https://news.daum.net/culture/">문화</a> <a class="link_txt" href="https://news.daum.net/digital/">IT</a> <a class="link_txt" href="https://news.daum.net/photo/">포토</a> <a class="link_txt" href="https://news.daum.net/tv/">TV</a> 
      </div> 
      <div class="item_bundle"> <a class="link_txt" href="https://news.daum.net/cp">언론사별 뉴스</a> 
      </div> 
      <div class="item_bundle"> <a class="link_txt" href="https://news.daum.net/newsbox">배열이력</a> 
      </div> 
      <div class="item_bundle"> <a class="link_txt" href="https://news.daum.net/breakingnews">전체뉴스</a> <a class="link_txt" href="https://news.daum.net/ranking/popular">랭킹</a> <a class="link_txt" href="https://news.daum.net/series">연재</a> <a class="link_txt" href="https://focus.daum.net/ch/news/factcheck">팩트체크</a> 
      </div> <button class="btn_fold" type="button"><span class="ico_news">바로가기 링크 더보기/접기</span></button> 
     </div> 
     <h2 class="screen_out">서비스 이용정보</h2> 
     <div class="footer_link"> <a class="link_info" href="https://news.daum.net/info/intro">다음뉴스 소개</a> <span class="txt_dot"> ・ </span> <a class="link_info" href="https://news.daum.net/notice">공지사항</a> <span class="txt_dot"> ・ </span> <a class="link_info" href="https://news.daum.net/info/bbsrule">게시물 운영원칙</a> <span class="txt_dot"> ・ </span> <a class="link_info" href="https://policy.daum.net/info/info">서비스 약관/정책</a> <span class="txt_dot"> ・ </span> <a class="link_info" href="https://with.kakao.com/media">뉴스제휴</a> <span class="txt_dot"> ・ </span> <a class="link_info" href="https://business.kakao.com/">비즈니스</a> <span class="txt_dot"> ・ </span> <a class="link_info" href="https://cs.daum.net/redbell/top.html">권리침해신고</a> <span class="txt_dot"> ・ </span> <a class="link_info" href="https://cs.daum.net/faq/63.html">다음뉴스 고객센터</a> 
     </div> 
     <div class="footer_link"> <a class="link_info" href="https://news.daum.net/info/newscenter24">24시간 뉴스센터</a> <span class="txt_dot"> ・ </span> <a class="link_info" href="https://news.daum.net/info/edit">서비스원칙</a> <span class="txt_dot"> ・ </span> <a class="link_info" href="https://news.daum.net/info/correct">바로잡습니다</a> <span class="txt_dot"> ・ </span> <span class="link_info">기사배열책임자 : 김희정</span> <span class="txt_dot"> ・ </span> <span class="link_info">청소년보호책임자 : 정현주</span> 
     </div> 
     <div class="info_copyright"> <strong class="info_warn">위 내용에 대한 저작권 및 법적 책임은 자료제공사 또는 <br>글쓴이에 있으며, Kakao의 입장과 다를 수 있습니다.</strong> <small class="txt_copyright">Copyright © <strong class="txt_kakao"><a href="https://www.kakaocorp.com" class="link_kakao">Kakao Corp.</a></strong> All rights reserved.</small> 
     </div> 
    </div>   
   </div> <samp data-cloud-area="util">           </samp> 
  <div>
 <div id="translatingLayer" class="translate_layer">
 <div class="inner_translate_layer">
 <div class="layer_body">
 <strong class="tit_translate">
 번역중 <span class="eng_tit">Now in translation</span>
 </strong>
 </div>
 </div>
 </div>

 <div id="translateErrorLayer" class="translate_layer">
 <div class="inner_translate_layer inner_error_layer">
 <div class="layer_body">
 <span class="ico_newsview ico_error"></span>
 <strong class="tit_translate">
 잠시 후 다시 시도해 주세요 <span class="eng_tit">Please try again in a moment</span>
 </strong>
 </div>
 </div>
 </div>
 </div></div>
```
